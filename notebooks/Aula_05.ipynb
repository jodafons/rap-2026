## Introdução a Strings: Criação e Indexação

Uma string é uma sequência de caracteres. Em Python, declaramos strings usando aspas simples (`'...'`), aspas duplas (`"..."`) ou aspas triplas (três aspas simples ou três aspas duplas, úteis para strings de múltiplas linhas).

**Importante:** Strings em Python são **imutáveis**. Isso significa que, uma vez criada, você não pode alterar um caractere individual da string diretamente (ex: tentar fazer `texto[0] = "A"` causará um erro).

---

### Strings como "Arrays" de Caracteres

Para compreender a indexação de strings, é fundamental deixar claro ao aluno que **uma string se comporta exatamente como um array (vetor) de caracteres**. Cada caractere ocupa uma posição específica na memória e pode ser acessado individualmente através de seu índice (index).

Veja o exemplo prático com a palavra **"SONAR"** mapeada no sistema de coordenadas de índices:

| Índice Positivo (Esquerda para Direita) | `0` | `1` | `2` | `3` | `4` |
| :--- | :---: | :---: | :---: | :---: | :---: |
| **Caractere** | **S** | **O** | **N** | **A** | **R** |
| **Índice Negativo (Direita para Esquerda)** | `-5` | `-4` | `-3` | `-2` | `-1` |

**Conceitos-chave para o aluno:**
1. A primeira letra da palavra `"SONAR"` está na posição `0`.
2. O índice **positivo** vai de `0` até o `comprimento da palavra - 1` (de esquerda para a direita).
3. O índice **negativo** vai de `-1` (último caractere) até `-comprimento da palavra` (de direita para a esquerda). É ideal para ler dados do fim para o começo!

```python
# 1. Declaração de strings com aspas simples e duplas
mensagem_simples = 'Radar ativo em funcionamento.'
alerta_sistema = "ATENÇÃO: Reator com temperatura elevada!"

print("Mensagem simples:", mensagem_simples)
print("Alerta de sistema:", alerta_sistema)
```

```python
# 2. Declaração de strings com aspas triplas (múltiplas linhas)
log_multilinha = """[LOG - 10:45:12] Sensor de profundidade ativado.
[LOG - 10:45:15] Pressão estável a 12 atm.
[LOG - 10:45:20] Nível de oxigênio: 98%."""

print("Log multilinha:\n", log_multilinha)
```

```python
# 1. Acesso a caracteres por índice (indexação positiva e negativa)
codigo_transmissao = "CIAA-LPS-992"

primeiro_caractere = codigo_transmissao[0]
ultimo_caractere = codigo_transmissao[-1]   # indexação negativa (último elemento)

print("Código original:", codigo_transmissao)
print("Primeiro caractere:", primeiro_caractere)
print("Último caractere:", ultimo_caractere)
```

```python
# 2. Fatiamento (slicing)
# Extraindo a sigla do submarino (índices de 0 a 7)
sigla_submarino = codigo_transmissao[0:8]
print("Sigla do Submarino (Slicing 0:8):", sigla_submarino)
```

```python
# 3. Tentativa de alteração (Imutabilidade)
# Remover o caractere '#' do comentário abaixo gerará um TypeError se executado:
# codigo_transmissao[0] = 'X' 

# Para alterar, precisamos criar uma nova string
codigo_corrigido = "X" + codigo_transmissao[1:]
print("Novo código criado a partir do fatiamento:", codigo_corrigido)
```

## Operações Básicas com Strings

Podemos realizar operações matemáticas simples e verificações de pertinência com strings:
*   `+` (Concatenação): Une duas strings.
*   `*` (Repetição): Repete a string N vezes.
*   `in` e `not in` (Operadores de pertinência): Verificam se uma substring está presente no texto.
*   `len(string)`: Retorna o número total de caracteres da string (incluindo espaços).

```python
# 1. Concatenação de strings (união de textos)
prefixo = "Submarino-"
numero_cauda = "C12"
nome_completo = prefixo + numero_cauda
print("Nome completo do contato:", nome_completo)
```

```python
# 2. Repetição de strings (multiplicação de textos)
divisor = "=" * 30
print(divisor)
print("     LOG DE OPERAÇÕES ACÚSTICAS")
print(divisor)
```

```python
# 1. Tamanho da string com len()
mensagem_radio = "Mensagem cifrada recebida do comando tático."
print(f"Tamanho da mensagem: {len(mensagem_radio)} caracteres.")
```

```python
# 2. Operadores de pertinência (in e not in)
contem_cifrada = "cifrada" in mensagem_radio
contem_hostil = "hostil" in mensagem_radio

print("A mensagem contém a palavra 'cifrada'?", contem_cifrada)
print("A mensagem contém a palavra 'hostil'?", contem_hostil)
```

## Métodos Essenciais de Modificação e Busca

O Python possui uma vasta coleção de métodos internos para strings. Veja a tabela resumo a seguir com os principais métodos que estudaremos nesta seção:

| Método | Descrição | Exemplo |
| :--- | :--- | :--- |
| `len(string)` | Retorna a quantidade de caracteres da string (função built-in). | `len("Sol") -> 3` |
| `.upper()` | Converte todos os caracteres para maiúsculas. | `"radar".upper() -> "RADAR"` |
| `.lower()` | Converte todos os caracteres para minúsculas. | `"SOM".lower() -> "som"` |
| `.title()` | Converte a primeira letra de cada palavra para maiúscula. | `"capitão lucas".title() -> "Capitão Lucas"` |
| `.strip()` | Remove espaços em branco no início e no fim da string. | `"  log  ".strip() -> "log"` |
| `.replace(antigo, novo)` | Substitui todas as ocorrências de um termo por outro. | `"A1".replace("A", "B") -> "B1"` |
| `.split(delimitador)` | Divide a string em uma lista de partes usando o delimitador. | `"10;20".split(";") -> ["10", "20"]` |
| `.join(lista)` | Une elementos de uma lista usando a string como separador. | `"-".join(["A", "B"]) -> "A-B"` |
| `.startswith(prefixo)` | Verifica se a string inicia com o prefixo (retorna True/False). | `"CIAA-9".startswith("CIAA") -> True` |
| `.endswith(sufixo)` | Verifica se a string termina com o sufixo (retorna True/False). | `"sonar.log".endswith(".log") -> True` |
| `.find(termo)` | Retorna a posição (índice) da primeira ocorrência do termo ou -1. | `"Alvo".find("v") -> 2` |

```python
# 1. Remoção de espaços extras com .strip()
comando_sujo = "   iniciar_propulsores_auxiliares   "
comando_limpo = comando_sujo.strip()

print(f"Original: '{comando_sujo}'")
print(f"Limpo com strip(): '{comando_limpo}'")
```

```python
# 2. Alteração de caixa (.upper(), .lower() e .title())
# Usamos a variável comando_limpo declarada na célula anterior
print("Maiúsculo (.upper()):", comando_limpo.upper())
print("Minúsculo (.lower()):", comando_limpo.lower())
print("Título (.title()):", "comandante lucas silva".title())
```

```python
# 2. Método .replace()
# Substitui todas as ocorrências de um termo por outro
comando_antigo = "status_valvula: FECHADA"
comando_novo = comando_antigo.replace("FECHADA", "ABERTA")

print("Original:", comando_antigo)
print("Após replace():", comando_novo)

# Exemplo de remoção de caracteres substituindo por string vazia ""
registro = "ID: #990-21"
registro_limpo = registro.replace("#", "")
print("Registro limpo:", registro_limpo)
```

```python
# 1. Divisão de string com .split()
dados_coordenadas = "-22.9083;-43.1964;150.5" # lat;lon;profundidade
elementos = dados_coordenadas.split(";")

print("Coordenadas originais:", dados_coordenadas)
print("Lista resultante do split(';'):", elementos)
```

```python
# 2. Junção de lista com .join()
lista_tripulantes = ["Silva", "Lucas", "Mariana"]
tripulacao_str = ", ".join(lista_tripulantes)

print("Lista original:", lista_tripulantes)
print("String unida por join():", tripulacao_str)
```

```python
# Exemplo de processamento de string com split e capitalização
# Recebemos uma mensagem de comando do sistema separada por sublinhados (underlines) '_'
comando_sistema = "alerta_de_pressao_na_cabine"

# 1. Fazemos o split por '_' para obter as palavras individuais
palavras = comando_sistema.split("_")
print("1. Palavras após o split('_'):", palavras)

# 2. Iteramos sobre a lista e aplicamos capitalize() em cada palavra
palavras_capitalizadas = []
for palavra in palavras:
    palavras_capitalizadas.append(palavra.capitalize())
print("2. Palavras com a primeira letra maiúscula:", palavras_capitalizadas)

# 3. Unimos novamente com espaços em branco usando .join()
mensagem_final = " ".join(palavras_capitalizadas)
print("3. Mensagem final formatada:", mensagem_final)
```

```python
# 1. Verificação de prefixo e sufixo (.startswith() e .endswith())
nome_arquivo = "telemetria_sonar_passivo.log"

eh_log = nome_arquivo.endswith(".log")
eh_telemetria = nome_arquivo.startswith("telemetria")

print(f"O arquivo '{nome_arquivo}' termina com '.log'?", eh_log)
print(f"O arquivo '{nome_arquivo}' começa com 'telemetria'?", eh_telemetria)
```

```python
# 2. Localização de substring com .find()
mensagem = "Alvo localizado no setor 3"
posicao = mensagem.find("setor")
print("Posição da palavra 'setor' na mensagem:", posicao)
```

## Exemplos Práticos Aplicados

Aqui veremos como combinar esses métodos para resolver tarefas táticas simuladas de bordo.

```python
# Exemplo A: Decodificador e Validador de Sinal de Rádio
# Recebemos um sinal contendo espaços extras, em minúsculas e com código de segurança
sinal_recebido = "   sinal_valido_CIAA-920-OK   "

def processar_sinal(sinal):
    # Limpa espaços nas pontas
    sinal_limpo = sinal.strip()
    # Converte para maiúsculo
    sinal_maiusculo = sinal_limpo.upper()
    # Verifica se começa com o prefixo do submarino
    if sinal_maiusculo.startswith("SINAL_VALIDO_CIAA"):
        # Extrai o código de segurança (tudo após o prefixo)
        codigo = sinal_maiusculo.replace("SINAL_VALIDO_CIAA-", "")
        return f"SINAL APROVADO! Código de segurança: {codigo}"
    else:
        return "SINAL ANORMAL: Acesso negado!"

resultado_sinal = processar_sinal(sinal_recebido)
print(resultado_sinal)
```

```python
# Exemplo B: Parser de Telemetria de Sensores
# Recebemos uma linha de dados no formato "sensor_nome | valor_medido | unidade"
telemetria_bruta = "profundidade_casco | 420.5 | metros"

def parse_telemetria(dado_bruto):
    # Divide os elementos pela barra vertical '|'
    partes = dado_bruto.split("|")
    
    # Limpa os espaços de cada parte usando strip e atribui às variáveis
    nome_sensor = partes[0].strip().upper()
    valor = float(partes[1].strip())
    unidade = partes[2].strip()
    
    return f"Sensor {nome_sensor} reportou leitura de {valor} {unidade}."

relatorio_sensor = parse_telemetria(telemetria_bruta)
print(relatorio_sensor)
```
