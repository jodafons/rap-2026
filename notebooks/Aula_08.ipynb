## Introdução a Sets

Um conjunto (`set`) é uma coleção de elementos que é **desordenada**, **mutável** e **não permite elementos duplicados**. 

Em Python, os conjuntos são representados por chaves `{}` (assim como os dicionários, mas contendo apenas valores soltos, sem o par `chave: valor`) ou através da função built-in `set()`.

### Para Que Servem os Sets?
1. **Remoção de Duplicatas**: É a forma mais rápida e eficiente em Python de remover elementos repetidos de uma lista ou tupla.
2. **Operações Matemáticas de Conjuntos**: Permite realizar operações clássicas da álgebra de conjuntos (união, interseção, diferença e diferença simétrica) com operadores simples.
3. **Testes de Pertinência de Alta Performance**: Verificar se um item está presente em um conjunto (`item in conjunto`) é executado em tempo constante ($O(1)$), de forma tão rápida quanto nos dicionários.

## Diferenças: Tupla, Set e Dicionário

Para entender quando usar cada estrutura de dados, veja a tabela comparativa:

| Estrutura | Sintaxe | Ordenado? | Acesso | Mutável? | Duplicados? | Caso de Uso Principal |
| :--- | :--- | :--- | :--- | :--- | :--- | :--- |
| **Tupla (`tuple`)** | `(1, 2)` | Sim | Índice numérico (`tup[0]`) | **Não (Imutável)** | Sim | Coordenadas fixas, registros de dados fixos. |
| **Set (`set`)** | `{1, 2}` | Não | Sem acesso direto (sem índice) | Sim | **Não** | Filtrar duplicatas, operações matemáticas. |
| **Dicionário (`dict`)** | `{"a": 1}`| Sim (Python 3.7+) | Chave customizada (`dict["a"]`) | Sim | Não (Chaves únicas) | Mapeamento chave-valor, cadastros. |

### Comparação de Detalhes
* **Tupla vs. Set**: Tuplas são imutáveis e ordenadas; permitem elementos repetidos e acesso por índice. Sets são mutáveis, desordenados, não permitem repetição e não suportam acesso por índice.
* **Set vs. Dicionário**: Ambos utilizam chaves `{}` e tabelas hash por baixo do pano para busca instantânea. No entanto, o dicionário armazena mapeamentos `chave: valor`, enquanto o set armazena apenas os valores únicos de forma isolada.

## Métodos e Operações de Sets

Abaixo estão listados os principais métodos e operadores usados para manipular conjuntos e realizar álgebra de conjuntos:

| Operação / Método | Descrição | Exemplo | Equivalente Gráfico / Operador |
| :--- | :--- | :--- | :--- |
| `.add(valor)` | Adiciona um elemento ao conjunto. | `conj.add(10)` | - |
| `.remove(valor)` | Remove o elemento especificado. Gera erro se não existir. | `conj.remove(5)` | - |
| `.discard(valor)`| Remove o elemento especificado. **Não** gera erro se não existir. | `conj.discard(5)` | - |
| `.union(outro)` | Retorna a união de dois conjuntos (todos os elementos). | `c1.union(c2)` | `c1 \| c2` |
| `.intersection(outro)`| Retorna os elementos presentes em ambos os conjuntos. | `c1.intersection(c2)` | `c1 & c2` |
| `.difference(outro)`| Retorna os elementos presentes apenas no primeiro conjunto. | `c1.difference(c2)` | `c1 - c2` |
| `.symmetric_difference()`| Retorna os elementos que estão em apenas um dos conjuntos. | `c1.symmetric_difference(c2)`| `c1 ^ c2` |

## Exemplos Básicos de Funcionamento

A seguir, veja os conceitos de conjuntos aplicados em células de código separadas.

```python
# 1. Criação de Sets
# Podemos criar usando chaves {} ou convertendo outras coleções com set()
sensores_ativos = {"sonar_leste", "sonar_oeste", "radar"}
sinais_repetidos = ["alvo_1", "alvo_2", "alvo_1", "alvo_3", "alvo_2"]

# Removendo duplicatas convertendo lista em set
sinais_unicos = set(sinais_repetidos)

# Atenção: para criar um conjunto VAZIO, você DEVE usar set(), pois {} cria um dicionário vazio!
conjunto_vazio = set()

print("Sensores Ativos:", sensores_ativos)
print("Sinais Únicos (sem duplicatas):", sinais_unicos)
print("Tipo de conjunto_vazio:", type(conjunto_vazio))
```

```python
# 2. Modificação de Conjuntos
frequencias = {450, 900}

# Adicionando elementos (se o elemento já existir, o set não muda)
frequencias.add(1200)
frequencias.add(900) # Duplicado, será ignorado

# Removendo elementos
frequencias.remove(900) # Remove e funciona
frequencias.discard(100) # Tenta remover 100. Como não existe, discard não gera erro (remove geraria)

print("Conjunto de frequências após modificações:", frequencias)
```

```python
# 3. Operações Matemáticas de Conjuntos
sonar_a = {"alvo_1", "alvo_2", "alvo_3"}
sonar_b = {"alvo_3", "alvo_4", "alvo_5"}

# União (|): todos os alvos detectados por pelo menos um sonar
todos_alvos = sonar_a | sonar_b

# Interseção (&): alvos comuns detectados por ambos os sonares
alvos_comuns = sonar_a & sonar_b

# Diferença (-): alvos detectados exclusivamente pelo sonar A
exclusivo_a = sonar_a - sonar_b

# Diferença Simétrica (^): alvos detectados por apenas um dos sonares (não por ambos)
nao_compartilhados = sonar_a ^ sonar_b

print("União (todos os alvos):", todos_alvos)
print("Interseção (comuns):", alvos_comuns)
print("Diferença (exclusivo do sonar A):", exclusivo_a)
print("Diferença Simétrica (não compartilhados):", nao_compartilhados)
```
