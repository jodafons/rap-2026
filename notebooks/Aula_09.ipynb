## Escopo de Variáveis (Local vs. Global)

O escopo determina a visibilidade e a "vida útil" de uma variável no seu programa.

* **Escopo Local**: Variáveis criadas **dentro** de uma função pertencem apenas a ela. Elas passam a existir quando a função começa a rodar e são destruídas da memória quando ela termina.
* **Escopo Global**: Variáveis criadas **fora** de todas as funções (no corpo principal do script) são visíveis por todo o programa.

### Regras importantes:
1. **Acesso**: Funções podem **ler** variáveis globais livremente.
2. **Modificação**: Por padrão, se você tentar alterar uma variável global dentro de uma função, o Python criará uma nova variável local com o mesmo nome, deixando a global de fora intacta.
3. **Palavra-chave `global`**: Se você precisar explicitamente alterar o valor de uma variável global de dentro de uma função, deverá declarar `global nome_da_variavel` antes de modificá-la. *(Use com muita cautela, pois abusar de globais dificulta o rastreamento de bugs!)*

```python
# 1. Escopo Local vs Global (Leitura)
identificacao_submarino = "CIAA-LPS-9" # Global

def exibir_dados():
    nivel_oxigenio = 98.5 # Local (só existe aqui dentro)
    print("[Dentro da Função] Lendo Oxigênio (local):", nivel_oxigenio)
    print("[Dentro da Função] Lendo Identificação (global):", identificacao_submarino)

exibir_dados()

# Tentativa de ler a variável local fora do seu escopo gerará um NameError
# print(nivel_oxigenio) # Descomente para ver o erro!
```

```python
# 2. Modificando Variáveis Globais (Uso de 'global')
frequência_sonar = 450 # Global

def alterar_frequencia_local(nova_freq):
    frequência_sonar = nova_freq # Cria uma variável LOCAL, não altera a global
    print("[alterar_frequencia_local] Frequência localizada:", frequência_sonar)

def alterar_frequencia_global(nova_freq):
    global frequência_sonar # Informa ao Python que modificaremos a variável global externa
    frequência_sonar = nova_freq
    print("[alterar_frequencia_global] Frequência global alterada:", frequência_sonar)

print("Antes das chamadas - Frequência Global:", frequência_sonar)
alterar_frequencia_local(600)
print("Após alterar_local - Frequência Global:", frequência_sonar)

alterar_frequencia_global(800)
print("Após alterar_global - Frequência Global:", frequência_sonar)
```

## Argumentos Padrão (Default Parameters)

Em Python, podemos definir valores padrões para os parâmetros de uma função. Se a função for chamada sem fornecer esses argumentos, os valores padrões serão usados automaticamente.

### Regra de Sintaxe de Argumentos:
> [!IMPORTANT]
> Parâmetros com valores padrão (opcionais) devem sempre ser declarados **depois** dos parâmetros sem valores padrão (obrigatórios).
>
> Exemplo correto: `def configurar(id_painel, brilho=80):`
> Exemplo incorreto: `def configurar(brilho=80, id_painel):` (gera erro de sintaxe)

```python
# 3. Argumentos Padrão em Ação
def enviar_alerta(alvo, nivel="ALERTA"):
    # Se o nível não for especificado, assume "ALERTA"
    print(f"[{nivel}] Alvo {alvo} detectado!")

# Chamada omitindo o argumento padrão
enviar_alerta("Submarino Hostil")

# Chamada fornecendo todos os argumentos
enviar_alerta("Cardume de Peixes", nivel="INFO")
```

## Empacotamento e Desempacotamento com *args e **kwargs

Às vezes, não sabemos de antemão quantos argumentos serão passados para a nossa função. O Python resolve isso com as sintaxes `*args` e `**kwargs`.

### `*args` (Argumentos Posicionais Variáveis)
* A sintaxe `*args` permite que uma função receba qualquer número de argumentos posicionais.
* Os argumentos passados são empacotados dentro de uma **Tupla (`tuple`)** chamada `args`.

### `**kwargs` (Argumentos Nomeados Variáveis)
* A sintaxe `**kwargs` (*keyword arguments*) permite que uma função receba qualquer quantidade de argumentos nomeados (`nome=valor`).
* Os argumentos são empacotados em um **Dicionário (`dict`)** chamado `kwargs`.

```python
# 4. Uso de *args (Empacotamento Posicional)
def listar_contatos(*alvos):
    # 'alvos' é recebido como uma tupla
    print("Tipo de 'alvos':", type(alvos))
    print("Coleção de alvos captados:")
    for alvo in alvos:
        print(f"- {alvo}")

# Podemos passar 1, 3 ou nenhum argumento
listar_contatos("ALVO-1")
print("-" * 30)
listar_contatos("ALVO-A", "ALVO-B", "ALVO-C")
```

```python
# 5. Uso de **kwargs (Empacotamento Nomeado)
def detalhar_status_reator(reator_id, **detalhes):
    # 'detalhes' é recebido como um dicionário
    print(f"Status do Reator: {reator_id}")
    for chave, valor in detalhes.items():
        print(f"- {chave.upper()}: {valor}")

# Passamos chaves-valores customizadas na chamada
detalhar_status_reator("REATOR-PRINCIPAL", temperatura=285.5, pressao=15.2, status="Normal")
```

## Tipagem de Dados em Funções (Type Hinting)

O Python é uma linguagem de tipagem dinâmica, o que significa que não precisamos declarar obrigatoriamente o tipo de uma variável ao criá-la. No entanto, em sistemas complexos (como a navegação do submarino), indicar quais tipos de dados uma função espera receber e retornar torna o código muito mais fácil de ler, documentar e depurar. 

Essa funcionalidade é chamada de **Type Hinting** (indicação de tipo).

### Tipos Simples vs. Estruturas Complexas

Para tipos de dados simples, usamos os próprios tipos built-in:
* `nome: str` (para texto)
* `valor: float` ou `quantidade: int`
* `ativo: bool`

Para estruturas de dados como listas, dicionários e tuplas, importamos utilitários adicionais do módulo nativo `typing` (com a primeira letra maiúscula):
* `List[tipo]` ou `list[tipo]`: Uma lista contendo elementos de determinado tipo.
* `Dict[tipo_chave, tipo_valor]` ou `dict[tipo_chave, tipo_valor]`: Um dicionário associando chaves e valores.
* `Tuple[tipo_1, tipo_2]` ou `tuple[tipo_1, tipo_2]`: Uma tupla contendo elementos nas posições especificadas.

*Nota: A partir do Python 3.9, você pode usar os tipos em letras minúsculas (`list`, `dict`, `tuple`) diretamente sem precisar importar o módulo `typing`, mas em projetos anteriores a importação com letras maiúsculas (`List`, `Dict`, `Tuple`) é o padrão.*

```python
# 6. Exemplo de Type Hinting
from typing import List, Dict, Tuple

# Esta função indica que recebe uma lista de floats, um dicionário com chave str e valor str,
# e retorna uma tupla contendo um float e um booleano.
def processar_telemetria(leituras: List[float], limites: Dict[str, float]) -> Tuple[float, bool]:
    soma = sum(leituras)
    limite_critico = limites.get("critico", 100.0)
    
    # Verifica se a soma ultrapassa o limite crítico
    sobrecarga = soma > limite_critico
    return soma, sobrecarga

# Utilizando a função normalmente
leituras_bateria = [35.5, 42.0, 15.0]
config_limites = {"alerta": 80.0, "critico": 90.0}

total, perigo = processar_telemetria(leituras_bateria, config_limites)
print(f"Total lido: {total} | Em perigo? {perigo}")
```

## Tabela Comparativa de Tipos de Parâmetros

| Parâmetro | Sintaxe na Função | Como os Dados São Recebidos | Exemplo de Chamada | Caso de Uso Principal |
| :--- | :--- | :--- | :--- | :--- |
| **Posicional Obrigatório** | `def f(x):` | Valor individual | `f(10)` | Parâmetros essenciais. |
| **Argumento Padrão (Default)**| `def f(x=5):` | Valor individual | `f()` ou `f(10)` | Parâmetros opcionais e configurações padrão. |
| **Argumentos Posicionais Variáveis (`*args`)** | `def f(*args):` | Tupla (`tuple`) | `f(1, 2, 3)` | Quando a função pode receber infinitos valores isolados (ex: somar tudo). |
| **Argumentos Nomeados Variáveis (`**kwargs`)**| `def f(**kwargs):`| Dicionário (`dict`)| `f(a=1, b=2)` | Passagem de parâmetros de configuração opcionais nomeados. |
