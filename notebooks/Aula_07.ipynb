## Mutabilidade vs Imutabilidade

Antes de entrarmos nas tuplas, é crucial entender um conceito que rege como o Python manipula dados na memória: a diferença entre objetos **mutáveis** (que podem mudar) e **imutáveis** (que nunca mudam).

No Python, cada variável que criamos aponta para um objeto armazenado na memória do computador.
* **Objetos Mutáveis (Mutable)**: São objetos cujos valores internos **podem ser alterados diretamente** na memória após a criação, sem que o objeto mude de endereço. Exemplos: `list` (listas) e `dict` (dicionários).
* **Objetos Imutáveis (Immutable)**: São objetos que **nunca podem ser alterados diretamente**. Se tentarmos "mudar" seu valor, o Python criará um novo objeto em um novo endereço de memória. Exemplos: `int`, `float`, `str` (strings), `bool` e `tuple` (tuplas).

### O Comportamento ao Passar Dados para Funções

Essa diferença é muito importante quando passamos variáveis como argumentos para funções:
1. **Passagem de Mutável**: Se passamos uma lista para uma função e ela a modifica (adicionando itens com `.append()`, por exemplo), a lista **original fora da função também será modificada** (efeito colateral).
2. **Passagem de Imutável**: Se passamos uma string ou número para uma função e tentamos "modificá-la" lá dentro, a variável externa **permanece intacta**, pois o Python protege o valor original criando uma cópia local.

```python
# 1. Comportamento com objetos MUTÁVEIS (Listas) em funções
def registrar_contato_radar(radar_lista, novo_contato):
    # A lista é mutável: o método append altera o objeto diretamente na memória
    radar_lista.append(novo_contato)
    print("[Dentro da Função] Lista de radar modificada:", radar_lista)

# Código principal
contatos_sonar = ["Submarino Alfa", "Baleia Jubarte"]
print("Antes da função:", contatos_sonar)

# Executamos a função passando a nossa lista
registrar_contato_radar(contatos_sonar, "Navio Cargueiro")

print("Depois da função (lista original modificada):", contatos_sonar)
```

```python
# 1.2 Comportamento com objetos MUTÁVEIS (Dicionários) em funções
def calibrar_sensor(dados_sensor, novo_valor_calibrado):
    # Dicionários também são mutáveis. Alterar ou adicionar chaves modifica o objeto na memória
    dados_sensor["valor"] = novo_valor_calibrado
    dados_sensor["calibrado"] = True
    print("[Dentro da Função] Dicionário do sensor modificado:", dados_sensor)

# Código principal
sensor_sonar = {
    "id": "SNS-992",
    "valor": 45.8,
    "calibrado": False
}
print("Antes da função:", sensor_sonar)

# Executamos a função passando o dicionário do sensor
calibrar_sensor(sensor_sonar, 48.2)

print("Depois da função (dicionário original modificado):", sensor_sonar)
```

```python
# 2. Comportamento com objetos IMUTÁVEIS (Strings) em funções
def atualizar_codigo_encriptacao(codigo_atual, nova_chave):
    # Strings são imutáveis. Fazer '+' cria um NOVO objeto e reatribui localmente
    codigo_atual = codigo_atual + "-" + nova_chave
    print("[Dentro da Função] Código local atualizado:", codigo_atual)
    return codigo_atual

# Código principal
sinal_submarino = "CIAA-LPS"
print("Antes da função:", sinal_submarino)

# Executamos a função passando a nossa string
codigo_retornado = atualizar_codigo_encriptacao(sinal_submarino, "SECURE")

print("Depois da função (sinal original intacto):", sinal_submarino)
print("Depois da função (retorno da função):", codigo_retornado)
```

## Introdução a Tuplas

Uma tupla é uma coleção de elementos que é **ordenada** e **imutável**. Em Python, as tuplas são representadas por parênteses `(...)` e seus elementos são separados por vírgula.

### Para Que Servem as Tuplas?
1. **Garantia de Integridade (Segurança)**: Se você tem um dado que nunca deve ser alterado (como a constante da velocidade do som na água ou a coordenada de um ponto de encontro), usar uma tupla garante que nenhuma parte do código alterará esse dado por acidente.
2. **Desempacotamento de Dados (Unpacking)**: Permite extrair facilmente elementos de coleções e atribuí-los a variáveis em uma única linha.
3. **Retornos Múltiplos de Funções**: É a forma padrão que o Python usa para retornar múltiplos valores de uma única função.

## Diferenças: Lista, Tupla e Dicionário

Para ajudar na escolha da estrutura de dados ideal para cada problema, veja a tabela comparativa:

| Estrutura | Sintaxe | Acesso | Mutável? | Elementos Duplicados? | Caso de Uso Principal |
| :--- | :--- | :--- | :--- | :--- | :--- |
| **Lista (`list`)** | `[1, 2]` | Índice numérico (`0, 1...`) | Sim (Mutável) | Sim | Histórico de sensores, filas e coleções dinâmicas. |
| **Tupla (`tuple`)** | `(1, 2)` | Índice numérico (`0, 1...`) | **Não (Imutável)** | Sim | Coordenadas fixas `(lat, lon)`, constantes e registros. |
| **Dicionário (`dict`)** | `{"a": 1}`| Chave customizada (`"a"`) | Sim (Mutável) | Não (Chaves únicas) | Mapeamento direto de IDs para registros, configurações. |

### Diferença entre Lista e Dicionário

Embora ambas sejam coleções mutáveis em Python, elas possuem propósitos e estruturas de dados bem diferentes:
* **Lista (`list`)**: É uma coleção ordenada de elementos acessados por sua **posição (índice numérico)**. É perfeita para manter sequências ordenadas ou históricos.
* **Dicionário (`dict`)**: É uma coleção de pares **chave-valor** acessada por uma **chave única personalizada** (como strings ou números). É ideal para modelar registros, tabelas de consulta rápida ou configurações de sistemas.

## Métodos e Funções de Tuplas

Por serem imutáveis, as tuplas possuem apenas dois métodos internos (além de funções built-in como `len()`):

| Função / Método | Descrição | Exemplo |
| :--- | :--- | :--- |
| `len(tupla)` | Retorna o número total de elementos na tupla. | `len((10, 20)) -> 2` |
| `.count(valor)` | Conta quantas vezes o valor especificado aparece na tupla. | `tup.count(5)` |
| `.index(valor)` | Retorna a posição (índice) da primeira ocorrência do valor especificado. Gera erro se não existir. | `tup.index("Alvo")` |

## Exemplos Básicos de Funcionamento

A seguir, veja os conceitos de tuplas aplicados em células de código separadas.

```python
# 1. Declaração de Tuplas
# Podemos declarar usando parênteses () ou apenas separando por vírgulas
coordenada_qg = (-22.9035, -43.1234) # Latitude e Longitude
limites_profundidade = (0, 150, 300, 450) # Metros

# Importante: para declarar uma tupla com um único elemento, é necessário usar uma vírgula no final!
tupla_unica = (100,) 

print("Coordenada QG:", coordenada_qg)
print("Tipo da variável coordenada_qg:", type(coordenada_qg))
print("Tupla com elemento único:", tupla_unica)
```

```python
# 2. Acesso e fatiamento (slicing)
# Funciona exatamente como nas listas e strings
print("Latitude QG (índice 0):", coordenada_qg[0])
print("Longitude QG (índice 1):", coordenada_qg[1])

# Fatiamento da tupla de limites
fatiado = limites_profundidade[1:3]
print("Fatiamento [1:3] (índices 1 e 2):", fatiado)
```

```python
# 3. Imutabilidade em ação
# Descomente a linha abaixo para ver que tentar modificar gera um TypeError:
# coordenada_qg[0] = -23.0000

print("As tuplas não permitem alterações de elementos após criadas!")
```

```python
# 4. Concatenação de Tuplas
# Embora as tuplas sejam imutáveis, podemos combiná-las usando o operador '+'
# Isso gera uma NOVA tupla contendo os elementos de ambas, sem alterar as originais!

coordenadas_leste = (10.0, 20.0)
coordenadas_oeste = (30.0, 40.0)

# Concatenando as duas tuplas
rota_completa = coordenadas_leste + coordenadas_oeste

print("Coordenadas Leste:", coordenadas_leste)
print("Coordenadas Oeste:", coordenadas_oeste)
print("Rota Completa (Concatenada):", rota_completa)

# Também podemos concatenar uma tupla com um elemento único (lembrando de usar a vírgula)
rota_completa_com_fim = rota_completa + (50.0,)
print("Rota Completa com Ponto Final:", rota_completa_com_fim)
```

```python
# 5. Desempacotamento de Tuplas (Unpacking)
# Permite atribuir os elementos da tupla a variáveis individuais em uma única linha
lat, lon = coordenada_qg

print("Latitude desempacotada:", lat)
print("Longitude desempacotada:", lon)
```

```python
# 6. Métodos .count() e .index()
mensagens_sonar = ("Submarino", "Baleia", "Submarino", "Ruído")

# Contando ocorrências de "Submarino"
total_submarinos = mensagens_sonar.count("Submarino")
print("Total de submarinos na tupla:", total_submarinos)

# Buscando o índice do primeiro "Baleia"
indice_baleia = mensagens_sonar.index("Baleia")
print("Posição da primeira baleia detectada:", indice_baleia)
```
