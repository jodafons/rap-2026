## Tipos Primitivos e Variáveis em Python

### Variáveis
Uma **variável** é um espaço na memória do computador reservado para armazenar dados. Em Python, a criação de uma variável ocorre no momento em que você atribui um valor a ela utilizando o operador `=`.

### Regras de Nomenclatura:
1. Deve começar com uma letra ou o caractere de sublinhado (`_`).
2. Não pode começar com números.
3. Só pode conter caracteres alfanuméricos e sublinhados (`A-z`, `0-9`, e `_`).
4. Nomes de variáveis fazem distinção entre maiúsculas e minúsculas (`idade`, `Idade` e `IDADE` são variáveis diferentes).

### Tipos Primitivos de Dados
O Python é dinamicamente tipado, ou seja, ele identifica automaticamente o tipo do dado. Os quatro tipos primitivos mais comuns são:
*   **Inteiros (`int`)**: Números inteiros, positivos ou negativos. Ex: `10`, `-5`.
*   **Ponto Flutuante (`float`)**: Números decimais. **Atenção**: usa-se ponto `.` e não vírgula. Ex: `3.14`, `-0.5`.
*   **Booleanos (`bool`)**: Valores lógicos: Verdadeiro (`True`) ou Falso (`False`).
*   **Strings (`str`)**: Textos delimitados por aspas simples ou duplas. Ex: `"Python"`, `'Programação'`.

### Casting (Conversão de Tipos)

O **Casting** é a conversão explícita de um tipo de dado para outro. Isso é feito por meio de funções construtoras:

*   `str()`: Converte um dado para texto (`str`).
*   `int()`: Converte para inteiro (`int`). Nota: ao converter um float, ele descarta a parte decimal (arredonda para baixo).
*   `float()`: Converte para número decimal (`float`).
*   `bool()`: Converte para booleano (`bool`). Em Python, `0`, strings vazias `""` e objetos vazios convertem para `False`, enquanto o resto converte para `True`.

Uma aplicação comum do casting é quando lemos valores do teclado com `input()`. Como a função `input()` sempre retorna uma string, precisamos convertê-la se quisermos efetuar cálculos matemáticos.

````python
# O sensor de bordo enviou o ano de comissão do reator como texto
ano_comissao_str = "2018"

# Convertemos de string para número para poder efetuar contas de idade
ano_comissao = int(ano_comissao_str)
ano_atual = 2026

anos_de_uso = ano_atual - ano_comissao
print(f"Reator ativo há {anos_de_uso} anos.")

# Exemplo de truncagem ao converter temperatura decimal
temperatura_sensor = 28.97
temperatura_inteira = int(temperatura_sensor)
print(f"Temperatura truncada do sensor: {temperatura_inteira}°C")
````

## Tipos de Prints e Formatações de Strings <br>

Exibir dados de maneira legível e bonita para o usuário final é essencial. Em Python, existem três formas principais de formatar textos com variáveis:

1. **Concatenação por vírgulas ou operador `+`** (Simples, mas pouco prático para textos longos).
2. **Estilo Antigo (Interpolador `%`)**: Utiliza marcadores como `%s` para strings, `%d` ou `%i` para inteiros e `%f` para floats.
3. **Método `.format()`**: Usa chaves `{}` como espaços reservados que são preenchidos na chamada do método.
4. **f-strings (Recomendado - Moderno)**: Introduzido no Python 3.6, permite colocar expressões diretamente dentro de chaves prefixando a string com `f`. É o método mais legível e rápido.

````python
nome_cadete = "Lucas"
nota_simulador = 9.7562
faltas_periodo = 2

# 1. Concatenação simples (vírgula adiciona espaço automático)
print("Cadete:", nome_cadete, "| Nota:", nota_simulador, "| Faltas:", faltas_periodo)

# 2. Estilo antigo (%)
# %.2f define duas casas decimais para o float, %s para string, %i para inteiro
print("Cadete: %s | Nota: %.2f | Faltas: %i (Estilo Antigo)" % (nome_cadete, nota_simulador, faltas_periodo))

# 3. Método .format()
print("Cadete: {} | Nota: {:.2f} | Faltas: {} (.format())".format(nome_cadete, nota_simulador, faltas_periodo))

# 4. f-strings (Recomendado - Mais legível)
print(f"Cadete: {nome_cadete} | Nota: {nota_simulador:.2f} | Faltas: {faltas_periodo} (f-string)")
````

## Controle de Fluxo: Estruturas Condicionais (`if`, `elif`, `else`)<br>

Na programação, muitas vezes precisamos que o computador tome decisões. Uma **estrutura condicional** avalia se uma expressão lógica é verdadeira (`True`) ou falsa (`False`), e direciona a execução do código de acordo com o resultado.

No Python, utilizamos três palavras-chave para isso:
*   `if` (se): Executa um bloco de código se a condição for verdadeira.
*   `elif` (senão se): Adiciona condições intermediárias caso o `if` seja falso. É a abreviação de *else if*.
*   `else` (senão): Executa um bloco de código de "salvaguarda" se nenhuma das condições anteriores for atendida.

### Operadores de Comparação Comuns:
*   `==` : Igual a
*   `!=` : Diferente de
*   `>`  : Maior que
*   `<`  : Menor que
*   `>=` : Maior ou igual a
*   `<=` : Menor ou igual a

### Operadores Lógicos Comuns:
*   `and` : Retorna `True` se ambas as condições forem verdadeiras.
*   `or`  : Retorna `True` se pelo menos uma das condições for verdadeira.
*   `not` : Inverte o valor lógico da condição (negação).


### Exemplo:



````python
# Exemplo A: Para acessar uma área restrita, o militar deve ser maior de idade (>= 18 anos) e 
# possuir autorização. Escreva um código para verificar se o militar está autorizado na sala 
# de baterias. Caso contrário, você deve negar. Caso seja menor de idade, bloqueio direto. 
# Caso não possua autorização, apenas, imprima uma mensagem onde a credencial é ausente 
# e que o mesmo deve solicitar liberação ao comando.

# Dados do tripulante que tenta acesso físico
idade_militar = 19
possui_autorizacao = True

print(f"Idade: {idade_militar} anos | Cartão de Acesso? {possui_autorizacao}")

# Verificando permissão de entrada usando condicionais
if idade_militar >= 18 and possui_autorizacao:
    print("Acesso Permitido: Compartimento das baterias liberado.")
elif idade_militar >= 18 and not possui_autorizacao:
    print("Acesso Negado: Credencial ausente. Solicite liberação ao Comando.")
else:
    print("Acesso Negado: Idade não regulamentar para zona de risco.")
````

## O que é uma Função? <br>

Uma **função** é um bloco de código estruturado e reutilizável que realiza uma tarefa específica. 

Imagine a função como um eletrodoméstico (por exemplo, um liquidificador):
1. **Entrada (Ingredientes)**: São as variáveis ou valores que enviamos para a função, chamados de **parâmetros** ou **argumentos**.
2. **Processamento**: É o trabalho que a função faz com os parâmetros dentro dela.
3. **Saída (Retorno)**: É o produto final gerado, que enviamos de volta ao programa principal usando a instrução `return`.

### A Regra da Indentação
Em Python, **não usamos chaves `{}`** para delimitar funções ou condicionais. Em vez disso, o escopo do código é definido pela **indentação** (o recuo do texto a partir da margem esquerda, geralmente com **4 espaços**). O cabeçalho da função termina com dois pontos `:` e todo o seu conteúdo interno deve estar recuado.

### Estrutura básica:
```python
def nome_da_funcao(parametro1, parametro2):
    # Código indentado pertencente à função
    resultado = parametro1 + parametro2
    return resultado
```

````python
# Exemplo 1: Função simples sem parâmetros e sem retorno
# História: O terminal operacional do submarino precisa exibir uma saudação
# padrão de inicialização de sistema sempre que o console é ligado.
def mensagem_boas_vindas():
    print("CIAA-LPS Submarine System v1.0")
    print("Pronto para operações acústicas táticas.")

# Chamando a função para ser executada:
mensagem_boas_vindas()
````

````python
# Exemplo 2: Função com parâmetros de entrada
# História: O console de rádio precisa registrar o nome do oficial de plantão 
# que assumiu o turno no submarino.
def saudar_aluno(nome_aluno, periodo):
    print(f"Oficial: {nome_aluno} | Turno: {periodo} | Status: Conectado.")

# Chamando a função enviando argumentos diferentes:
saudar_aluno("Ana", "Manhã")
saudar_aluno("Roberto", "Noite")
````

````python
# Example 3: Função com parâmetros e retorno de valor (return)
# História: O receptor de sonar captou uma leitura acústica e precisa 
# amplificar matematicamente a amplitude do sinal em duas vezes.
def calcular_dobro(numero):
    dobro = numero * 2
    return dobro

# Chamando a função e salvando a resposta em uma variável
sinal_amplificado = calcular_dobro(8.5)
print(f"Amplitude original: 8.5 dB | Amplificada: {sinal_amplificado} dB")
````

### Exemplo A: Conversor de Temperatura<br>

**O Problema**: Desenvolver um programa que converta uma temperatura dada na escala Celsius para a escala Fahrenheit.

**Explicação do Algoritmo**:
Este algoritmo realiza uma conversão matemática linear baseada na proporção de variação entre duas escalas termométricas. A escala Celsius fixa a fusão do gelo em $0^\circ\text{C}$ e a ebulição da água em $100^\circ\text{C}$ (intervalo de 100 graus). A escala Fahrenheit fixa os mesmos fenômenos em $32^\circ\text{F}$ e $212^\circ\text{F}$, respectivamente (intervalo de 180 graus).

A relação matemática de equivalência é dada pela fórmula:
$$F = \left(C \times \frac{9}{5}\right) + 32$$

Onde:
1. Multiplicamos a entrada em Celsius pela razão $\frac{9}{5}$ (ou $1.8$) para ajustar a escala de grandeza.
2. Somamos $32$ para deslocar a origem comum dos termômetros.
3. Exibimos a temperatura resultante formatada com apenas uma casa decimal utilizando **f-strings**.

````python
def celsius_para_fahrenheit(celsius):
    # Aplicação direta da fórmula matemática
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

# Executando e testando o algoritmo
temp_c = 26.5
temp_f = celsius_para_fahrenheit(temp_c)

print(f"{temp_c:.1f}°C equivalem a {temp_f:.1f}°F.")
````

### Exemplo B: Verificador de Paridade (Par ou Ímpar)<br>

**O Problema**: Criar uma função capaz de receber um número inteiro e identificar se ele é classificado como "Par" ou "Ímpar".

**Explicação do Algoritmo**:
Por definição matemática, um número inteiro é considerado **par** se for divisível de maneira exata por $2$. Ou seja, quando dividimos o número por 2, o resto da divisão deve ser exatamente zero.

Em computação, para extrair o resto de uma divisão inteira, utilizamos o operador de módulo `%`. 

A lógica do algoritmo segue estes passos:
1. O número é recebido e testado na condição: `numero % 2 == 0`.
2. Se a condição for verdadeira, a função interrompe a execução e retorna o texto `"Par"`.
3. Se a condição for falsa (o resto foi 1), o bloco `else` é acionado e a função retorna `"Ímpar"`.
4. A indentação garante que as instruções de retorno pertençam aos respectivos caminhos da estrutura condicional `if-else`.

````python
def verificar_paridade(numero):
    # Testa se o resto da divisão por 2 é zero
    if numero % 2 == 0:
        return "Par"
    else:
        return "Ímpar"

# Executando e testando o algoritmo
num1 = 44
num2 = 13

print(f"O número {num1} é {verificar_paridade(num1)}.")
print(f"O número {num2} é {verificar_paridade(num2)}.")
````

### Exemplo C: Calculadora de IMC (Índice de Massa Corporal)<br>

**O Problema**: Desenvolver um programa que receba o peso e a altura de uma pessoa, efetue o cálculo do IMC e forneça o diagnóstico de saúde com base nos intervalos oficiais recomendados.

**Explicação do Algoritmo**:
O Índice de Massa Corporal (IMC) é uma métrica internacional adotada pela OMS (Organização Mundial da Saúde). O cálculo consiste em relacionar o peso em quilogramas com a altura elevada ao quadrado:
$$IMC = \frac{peso}{altura^2}$$

Após calcular o IMC numérico, o algoritmo executa uma estrutura de controle condicional múltipla e sequencial (`if`, `elif`, `else`) para categorizar o resultado de acordo com as seguintes faixas:
*   **Menor que 18.5**: Abaixo do peso ideal.
*   **Entre 18.5 e 24.9**: Peso considerado normal/adequado.
*   **Entre 25.0 e 29.9**: Sobrepeso.
*   **Maior ou igual a 30.0**: Obesidade.

Ao final, o programa desempacota o retorno da função e exibe os resultados formatados de forma clara para o usuário.

````python
def calcular_distancia_sonar(velocidade_som, tempo_eco):
    # O tempo do eco representa o trajeto de ida e volta, dividimos por 2
    distancia = (velocidade_som * tempo_eco) / 2
    return distancia

def relatorio_de_combate(frequencia, amplitude, velocidade_som, tempo_eco):
    # Chamamos a função auxiliar que calcula a distância (retorna um único float)
    distancia = calcular_distancia_sonar(velocidade_som, tempo_eco)
    
    # Classificação tática baseada em condicionais
    if 500 <= frequencia <= 1500 and amplitude >= 80:
        alvo = "Submarino"
        print(f"🔥 ALERTA TÁTICO: Submarino hostil detectado a {distancia:.1f} metros de distância!")
    elif 20 <= frequencia <= 200 and amplitude >= 60:
        alvo = "Baleia"
        print(f"🐋 CONTATO BIOLÓGICO: Baleia avistada a {distancia:.1f} metros.")
    else:
        alvo = "Ruído"
        print(f"Log Sonar: Apenas ruídos naturais do oceano a {distancia:.1f} metros.")

# Executando e testando o algoritmo
velocidade_agua = 1500.0
relatorio_de_combate(850.0, 95.0, velocidade_agua, 1.4)
````

````python
# IMPLEMENTAÇÃO DO EXERCÍCIO 1

def analisar_sensores(temperatura, pressao, oxigenio):
    # Avaliando as regras de segurança sequencialmente
    if temperatura < 18.0 or temperatura > 28.0:
        return "ALERTA_TEMPERATURA"
    elif pressao < 0.9 or pressao > 1.1:
        return "ALERTA_PRESSAO"
    elif oxigenio < 19.0:
        return "ALERTA_OXIGENIO"
    else:
        return "SISTEMA_OK"

def painel_controle_seguranca(temperatura, pressao, oxigenio):
    # Chamando a função auxiliar e capturando o retorno em formato string
    codigo_alerta = analisar_sensores(temperatura, pressao, oxigenio)
    
    # Tomando decisões com base no status de retorno
    if codigo_alerta == "SISTEMA_OK":
        print(f"[Painel] Temperatura: {temperatura}°C | Pressão: {pressao} atm | O2: {oxigenio}%")
        print(f"Log: Condições da cabine seguras ({codigo_alerta}). Tripulação estável.\n")
    else:
        print("[⚠️ ALARME ACIONADO ⚠️]")
        print(f"Alerta Crítico detectado! Código de falha: {codigo_alerta}")
        print(f"Valores atuais -> Temp: {temperatura}°C | Pressão: {pressao} atm | O2: {oxigenio}%\n")

# --- Testando os cenários ---
# Cenário 1: Tudo ideal
painel_controle_seguranca(23.5, 1.0, 20.8)

# Cenário 2: Baixo oxigênio
painel_controle_seguranca(22.0, 1.05, 17.5)
````

### Exercício 2: Monitor de Sonar e Classificação de Alvos<br>

**O Desafio**:
O operador de sonar acústico registrou ondas sonoras na água. Para identificar se o sinal reflete uma baleia (biologia), um submarino hostil (ameaça) ou apenas ruído oceânico, o computador tático precisa calcular a distância e avaliar a classificação do alvo.

Você deve criar duas funções:
1.  **`calcular_distancia_sonar(velocidade_som, tempo_eco)`** (Função Auxiliar):
    *   **Entradas (2 parâmetros)**: `velocidade_som` (float em m/s) e `tempo_eco` (float em s).
    *   **Saídas (1 retorno)**: Um float indicando a distância calculada até o alvo (em metros).
    *   *Fórmula da Distância*: O eco do sonar demorou o `tempo_eco` para ir e voltar até o alvo. Sabendo que a distância percorrida é de ida e volta, a fórmula é:
        $$\text{distancia} = \frac{\text{velocidade\_som} \times \text{tempo\_eco}}{2}$$
2.  **`relatorio_de_combate(frequencia, amplitude, velocidade_som, tempo_eco)`** (Função Principal):
    *   Deve chamar `calcular_distancia_sonar` com os argumentos recebidos para obter a distância.
    *   Usando estruturas de controle (`if/elif/else`), deve classificar o alvo:
        *   Se a frequência for entre $500\text{ Hz}$ e $1500\text{ Hz}$ e a amplitude for maior ou igual a $80\text{ dB}$, classifique como `'Submarino'` e exiba:
            `"🔥 ALERTA TÁTICO: Submarino hostil detectado a [DISTÂNCIA] metros de distância!"`
        *   Se a frequência for entre $20\text{ Hz}$ e $200\text{ Hz}$ e a amplitude for maior ou igual a $60\text{ dB}$, classifique como `'Baleia'` e exiba:
            `"🐋 CONTATO BIOLÓGICO: Baleia avistada a [DISTÂNCIA] metros."`
        *   Caso contrário, classifique como `'Ruído'` e exiba:
            `"Log Sonar: Apenas ruídos naturais do oceano a [DISTÂNCIA] metros."`

````python
# IMPLEMENTAÇÃO DO EXERCÍCIO 2

def calcular_distancia_sonar(velocidade_som, tempo_eco):
    distancia = (velocidade_som * tempo_eco) / 2
    return distancia

def relatorio_de_combate(frequencia, amplitude, velocidade_som, tempo_eco):
    # Chamando a função auxiliar para calcular a distância
    dist = calcular_distancia_sonar(velocidade_som, tempo_eco)
    
    # Classificação tática com base em condicionais
    if 500 <= frequencia <= 1500 and amplitude >= 80:
        print(f"🔥 ALERTA TÁTICO: Submarino hostil detectado a {dist:.1f} metros de distância!")
    elif 20 <= frequencia <= 200 and amplitude >= 60:
        print(f"🐋 CONTATO BIOLÓGICO: Baleia avistada a {dist:.1f} metros.")
    else:
        print(f"Log Sonar: Apenas ruídos naturais do oceano a {dist:.1f} metros.")

# --- Testando a leitura acústica ---
velocidade_media_agua = 1500.0 # m/s

# Caso 1: Frequência média alta com muita amplitude
print("Sensor Sonar - Leitura 01:")
relatorio_de_combate(850.0, 95.0, velocidade_media_agua, 1.4)

# Caso 2: Canto de baleia detectado
print("\nSensor Sonar - Leitura 02:")
relatorio_de_combate(50.0, 72.0, velocidade_media_agua, 1.2)
````
