## Introdução às Estruturas de Repetição (Loops)

Na programação, as **estruturas de repetição** (também chamadas de *laços* ou *loops*) permitem que um bloco de código seja executado repetidas vezes até que uma condição específica de parada seja alcançada ou enquanto percorremos uma sequência.

Imagine a necessidade de processar 1.000 leituras de um hidrofone de sonar. Escrever a linha de código 1.000 vezes seria impraticável. Com estruturas de repetição, escrevemos a instrução apenas uma vez e instruímos o computador a executá-la para cada leitura.

No Python, temos dois tipos principais de loops:

*   **Loop `for`**: Ideal para iterar sobre coleções ou quando sabemos previamente a quantidade de vezes que o código deve ser repetido (ou quando queremos passar por todos os itens de uma sequência, como uma string ou um intervalo de números).
*   **Loop `while`**: Ideal para quando não sabemos exatamente quantas vezes a repetição vai acontecer, pois ela continua rodando **enquanto** uma condição lógica for verdadeira. A condição deve se tornar falsa em algum momento para evitar o temido 'loop infinito'.

## Exemplos com Funções e Estrutura de Repetição `for`

O loop `for` em Python é extremamente poderoso porque ele itera diretamente sobre os elementos de uma sequência (como caracteres de um texto ou números em uma sequência).

### A Função `range()`
Uma ferramenta comum usada com o `for` é a função integrada `range()`. Ela gera uma sequência de números inteiros ajustável:
*   `range(parada)`: Gera números de 0 até o valor anterior à `parada`.
*   `range(inicio, parada)`: Começa no `inicio` e vai até antes da `parada`.
*   `range(inicio, parada, passo)`: Incrementa os números de acordo com o `passo`.

Vejamos exemplos práticos de sua aplicação dentro de funções:

````python
# Exemplo A: Gerador de Contagem Regressiva para Lançamento de Sonda
def contagem_regressiva(segundos):
    print(f"Iniciando sequência de contagem regressiva a partir de {segundos} segundos:")
    # range(inicio, parada, passo) -> vai de segundos até 1, diminuindo de 1 em 1
    for i in range(segundos, 0, -1):
        print(f"T-minus {i}...")
    print("Lançamento efetuado com sucesso!\n")

# Executando a função
contagem_regressiva(5)
````

````python
# Exemplo B: Simulador de Emissões de Pings do Sonar
def emitir_pings(total_pings):
    print(f"Transmitindo sequência de {total_pings} pings acústicos:")
    # O loop executa de 1 até total_pings
    for ping in range(1, total_pings + 1):
        # Amplitude fictícia calculada por fórmula matemática
        amplitude = ping * 1.5
        print(f"-> Ping #{ping} transmitido com amplitude {amplitude} dB.")
    print()

# Chamando a função
emitir_pings(4)
````

### Iterando com Índices usando a Função `enumerate()`

Muitas vezes, além de processar o valor de um caractere em uma string de texto em um loop `for`, precisamos saber a sua **posição** (o índice).

A função integrada `enumerate()` nos ajuda exatamente nisso. Ela retorna, a cada iteração do loop, um par contendo o índice (por padrão começando em 0) e a letra correspondente.

**Sintaxe Básica:**
```python
for indice, letra in enumerate(texto):
    # bloco de código utilizando o índice e o caractere
```

````python
# Exemplo C: Varredura de Letras do Sinal de Código SOS usando enumerate()
def analisar_sinal_codigo(codigo_str):
    print(f"Analisando caracteres da transmissão acústica '{codigo_str}':")
    # O enumerate nos devolve a posição do caractere (indice) e a respectiva letra
    for posicao, caractere in enumerate(codigo_str):
        print(f"Posição #{posicao}: caractere '{caractere}'")
    print()

# Analisando a mensagem de socorro
analisar_sinal_codigo("SOS")
````

## Exemplos com Funções e Estrutura de Repetição `while`

O loop `while` testa a condição antes de executar o bloco interno. Se for verdadeira, executa. Ao final do bloco, testa novamente. 

**Atenção**: É fundamental que a variável testada na condição seja atualizada dentro do bloco de código do loop, caso contrário o programa ficará preso executando para sempre.

````python
# Exemplo A: Simulação de Controle de Profundidade de Subida
def simular_subida(profundidade_inicial, velocidade_subida):
    profundidade_atual = profundidade_inicial
    segundos = 0
    
    print(f"Iniciando subida controlada. Profundidade inicial: {profundidade_inicial} metros.")
    
    # Repete enquanto o submarino não atingir a superfície (profundidade maior que 0)
    while profundidade_atual > 0:
        print(f"Tempo: {segundos}s | Profundidade: {profundidade_atual:.1f}m")
        
        # Atualiza a profundidade e o tempo
        profundidade_atual -= velocidade_subida
        segundos += 1
        
    print(f"Superfície alcançada! Tempo total de subida: {segundos} segundos.\n")

# Executando a simulação
simular_subida(15.0, 3.5)
````

## Controles de Fluxo `break` e `continue`

Em cenários reais, precisamos ter flexibilidade sobre a execução de um loop. O Python nos fornece duas ferramentas cruciais de controle:

### O Comando `break`
O `break` interrompe a execução do loop imediatamente e salta para a primeira linha de código fora do bloco de repetição. É frequentemente usado quando encontramos o que estávamos procurando e não há necessidade de continuar a busca.

### O Comando `continue`
O `continue` encerra apenas a iteração atual do loop. O interpretador ignora as linhas de código restantes abaixo do `continue` e salta diretamente para o teste de condição da próxima iteração. É ideal para ignorar elementos indesejados ou ruídos durante um processamento.

````python
# Visualizando o funcionamento do break
print("--- Exemplo do Break ---")
for numero in range(1, 6):
    if numero == 4:
        print(f"Encontrado {numero}. Parando o loop!")
        break
    print(f"Número: {numero}")

# Visualizando o funcionamento do continue
print("\n--- Exemplo do Continue ---")
for numero in range(1, 6):
    if numero == 4:
        print(f"Ignorando o número {numero}.")
        continue
    print(f"Número: {numero}")
````

## Exemplos Práticos Combinados de Repetição e Controle de Fluxo

Vejamos agora dois algoritmos práticos que demonstram como esses conceitos são estruturados em sistemas reais.

### Exemplo A: Varredura Tática de Canais de Rádio

**Algoritmo**:
Buscamos identificar qual canal específico de rádio está transmitindo um sinal hostil. Escaneamos sequencialmente de um canal inicial a um canal final. Assim que identificamos o canal que possui a assinatura hostil, interrompemos imediatamente a varredura (`break`), otimizando o tempo.

````python
def escanear_frequencias(canal_inicial, canal_final, canal_alvo):
    print(f"Iniciando varredura tática de canais do rádio:")
    for canal in range(canal_inicial, canal_final + 1):
        print(f"Escaneando canal #{canal}...")
        if canal == canal_alvo:
            print(f"🔥 SINAL HOSTIL DETECTADO no canal #{canal}! Interrompendo varredura.\n")
            break

# Escaneando canais de 10 a 15, com canal hostil no 12
escanear_frequencias(10, 15, 12)
````

### Exemplo B: Filtragem de Canais de Transmissão Restritos

**Algoritmo**:
Precisamos testar transmissões de sinal em canais sequenciais de 1 a N. No entanto, alguns canais específicos (por exemplo, 3 e 5) são reservados estritamente para o Alto Comando da Marinha e não devem receber sinais de teste. O algoritmo usa `continue` para ignorar esses canais e testar todos os demais normalmente.

````python
def testar_canais_transmissao(total_canais):
    print(f"Iniciando testes de transmissão nos canais (1 a {total_canais}):")
    for canal in range(1, total_canais + 1):
        if canal == 3 or canal == 5:
            # Canais de uso restrito do Comando. Pular teste.
            print(f"[!] Canal #{canal} reservado. Teste pulado.")
            continue
        print(f"[✓] Canal #{canal}: Sinal de teste transmitido com sucesso.")
    print()

# Testando canais de 1 a 6
testar_canais_transmissao(6)
````

### Exercício 1: Somatório de Números Pares

**O Objetivo**: Criar uma lógica simples de loop para somar apenas os números pares dentro de um intervalo.

**Instruções**:
Escreva uma função chamada `somar_pares(limite_maximo)` que:
1.  Inicialize um acumulador chamado `soma` em 0.
2.  Utilize o loop `for` e a função `range()` para percorrer todos os números inteiros de 1 até o `limite_maximo` (inclusive).
3.  Dentro do loop, verifique se o número atual é par (resto da divisão por 2 igual a zero).
4.  Se for par, adicione o valor ao acumulador `soma`.
5.  Retorne o valor final da soma.

````python
# IMPLEMENTAÇÃO DO EXERCÍCIO 1

def somar_pares(limite_maximo):
    soma = 0
    # O range deve ir até limite_maximo + 1 para incluir o limite_maximo na iteração
    for i in range(1, limite_maximo + 1):
        if i % 2 == 0:
            soma += i
    return soma

# Testando a função com limite 10 (soma esperada: 2 + 4 + 6 + 8 + 10 = 30)
resultado_soma = somar_pares(10)
print(f"Soma dos números pares até 10: {resultado_soma}")
````

### Exercício 2: Contagem de Ocorrências em Mensagem

**O Objetivo**: Iterar diretamente sobre uma string (cadeia de caracteres) usando o `for` para contar ocorrências de uma determinada letra.

**Instruções**:
Crie uma função chamada `contar_caracteres(texto, caractere_procurado)` que:
1.  Inicialize um contador em 0.
2.  Use um loop `for` para passar por cada caractere individual do `texto`.
3.  Verifique se o caractere atual é exatamente igual ao `caractere_procurado` (lembre-se que diferencia maiúsculas de minúsculas).
4.  Se for igual, aumente o contador em 1.
5.  Ao final, retorne a quantidade encontrada.

````python
# IMPLEMENTAÇÃO DO EXERCÍCIO 2

def contar_caracteres(texto, caractere_procurado):
    contador = 0
    for letra in texto:
        if letra == caractere_procurado:
            contador += 1
    return contador

# Testando a função
mensagem_teste = "SINAL ACUSTICO DETECTADO NO CANAL NORTE"
ocorrencias = contar_caracteres(mensagem_teste, "A")
print(f"A letra 'A' aparece {ocorrencias} vezes na mensagem.")
````

### Exercício 3: Varredura de Frequência de Emergência

**História e Enredo**:
O submarino está navegando em águas hostis quando de repente capta um sinal fraco. O operador de rádio precisa fazer uma varredura sequencial de frequências em busca da assinatura tática de comunicação aliada para decodificar um alerta.

**Instruções do Exercício**:
Crie uma função principal chamada `varrer_frequencias(limite_inicial, limite_final, frequencia_alvo)` que execute as seguintes tarefas:
1.  Utilize um loop `while` para incrementar as frequências de varredura de $10\text{ em }10\text{ Hz}$, iniciando em `limite_inicial`.
2.  A cada iteração, imprima no console a frequência sob análise (ex: `'Varrendo: 230 Hz'`).
3.  Se a frequência de varredura atingir a `frequencia_alvo` com precisão, imprima a mensagem: `'Frequência de emergência encontrada!'` e interrompa imediatamente o loop utilizando o comando `break`.
4.  Se a varredura atingir o `limite_final` sem encontrar nada, a repetição deve terminar normalmente.
5.  Ao final, o programa deve retornar a frequência em que a detecção ocorreu (ou `-1` se o loop for concluído sem sucesso).

````python
# IMPLEMENTAÇÃO DO EXERCÍCIO 3

def varrer_frequencias(limite_inicial, limite_final, frequencia_alvo):
    frequencia_atual = limite_inicial
    alvo_encontrado_freq = -1
    
    while frequencia_atual <= limite_final:
        print(f"Varrendo: {frequencia_atual} Hz")
        
        if frequencia_atual == frequencia_alvo:
            print("Frequência de emergência encontrada!")
            alvo_encontrado_freq = frequencia_atual
            break
            
        frequencia_atual += 10
        
    return alvo_encontrado_freq

# Testando a função com um sinal simulado
frequencia_detectada = varrer_frequencias(100, 300, 180)
print(f"Frequência identificada de emergência: {frequencia_detectada} Hz")
````

### Exercício 4: Processador de Pings Acústicos Sequenciais

**História e Enredo**:
O sensor de sonar do submarino está registrando as amplitudes dos ecos acústicos gerados por uma sequência matemática para os próximos `total_pings`. No entanto, devido à reverberação das ondas, alguns pings representam ruído nulo ou negativo e devem ser pulados (`continue`). Apenas pings positivos saudáveis devem ser acumulados à energia líquida total.

**Instruções do Exercício**:
Crie uma função chamada `processar_energia_pings(total_pings)`:
1.  Inicialize a variável `energia_total` em 0.0.
2.  Utilize um loop `for` e `range()` de 1 até `total_pings` (inclusive).
3.  Calcule a amplitude simulada do ping atual `i` usando a fórmula: 
    $$\text{amplitude} = (i \times 3) - 10$$
4.  Se a `amplitude` for menor ou igual a 0.0, imprima: `'Ping #[i] ignorado (ruído).'` e salte para a próxima iteração imediatamente usando o comando `continue`.
5.  Caso contrário, adicione a `amplitude` ao acumulador `energia_total` e exiba a mensagem: `'Ping #[i] processado com amplitude de [amplitude] dB.'`.
6.  Ao final do processamento, a função deve retornar o valor acumulado da `energia_total` arredondado para duas casas decimais com a função `round()`.

````python
# IMPLEMENTAÇÃO DO EXERCÍCIO 4

def processar_energia_pings(total_pings):
    energia_total = 0.0
    
    for i in range(1, total_pings + 1):
        amplitude = (i * 3) - 10
        
        if amplitude <= 0.0:
            print(f"Ping #{i} ignorado (ruído).")
            continue
            
        energia_total += amplitude
        print(f"Ping #{i} processado com amplitude de {amplitude} dB.")
        
    return round(energia_total, 2)

# Processando 6 pings simulados pela fórmula acústica
total_energia_limpa = processar_energia_pings(6)
print(f"\nEnergia total acumulada limpa: {total_energia_limpa} dB")
````
