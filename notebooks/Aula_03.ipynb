## 1) Exemplo Prático: Loops for, if/else e Funções

Antes de avançarmos para as listas, vamos consolidar a integração de loops, condicionais e funções.

Imagine que precisamos simular o monitoramento da pressão da câmara de ar comprimido para o sistema de lastro do submarino por um determinado número de ciclos. A cada ciclo de teste, realizamos uma leitura de pressão simulada. Se a leitura for superior a um limite seguro de tolerância, devemos emitir um alerta crítico. Caso contrário, reportamos que o sistema está operando estavelmente.

#### 📝 Roteiro Lógico (Pseudocódigo)
```text
Função monitorar_pressao_lastro(total_ciclos, limite_segurança):
    Escrever mensagem de início de monitoramento
    
    Para ciclo variando de 1 até total_ciclos (inclusive):
        pressao_lida = 1.0 + (ciclo * 0.05) - (resto de (ciclo / 3) * 0.08)
        
        Se pressao_lida > limite_segurança:
            Escrever alerta de pressão crítica
        Senão:
            Escrever log de pressão estável
            
    Escrever mensagem de conclusão
```

````python
def monitorar_pressao_lastro(total_ciclos, limite_segurança):
    """
    Simula o monitoramento da pressão de ar de lastro por N ciclos e emite alertas.
    """
    print(f"=== INICIANDO MONITORAMENTO DE SEGURANÇA ({total_ciclos} ciclos) ===")
    
    for ciclo in range(1, total_ciclos + 1):
        # Simulação matemática da pressão variando em função do ciclo
        pressao_lida = 1.0 + (ciclo * 0.05) - (ciclo % 3 * 0.08)
        
        if pressao_lida > limite_segurança:
            print(f"Ciclo #{ciclo}: Pressão = {pressao_lida:.2f} atm | ⚠️ ALERTA: Pressão crítica excedida!")
        else:
            print(f"Ciclo #{ciclo}: Pressão = {pressao_lida:.2f} atm | [OK] Pressão estável.")
            
    print("=== MONITORAMENTO CONCLUÍDO ===\n")

# Executando a simulação de monitoramento para 6 ciclos com limite de 1.15 atm
monitorar_pressao_lastro(total_ciclos=6, limite_segurança=1.15)
````

## 2) Exemplo Prático: Iteração Avançada com enumerate(), break e continue

A função integrada `enumerate()` é ideal para quando precisamos iterar sobre uma sequência (como os caracteres de uma mensagem codificada) e obter simultaneamente a **posição (índice)** e o **valor** do elemento atual.

Para controlar o comportamento do loop, relembramos o uso de:
*   `continue`: Interrompe a iteração atual do loop e pula imediatamente para a próxima.
*   `break`: Interrompe o loop por completo, saindo de sua execução imediatamente.

**Exemplo Aplicado**:
O operador do rádio do submarino interceptou uma cadeia de sinais codificados como uma string de texto. Cada caractere representa uma leitura acústica específica:
*   `'S'` (Safe): Sinal seguro (deve ser processado).
*   `'R'` (Ruído): Ruído de fundo (deve ser ignorado usando `continue`).
*   `'H'` (Hostile): Sinal de ameaça hostil detectado (o escaneamento deve ser imediatamente abortado por segurança usando `break`).

#### 📝 Roteiro Lógico (Pseudocódigo)
```text
Função escanear_transmissao_cripto(mensagem_codificada):
    Escrever mensagem de início de escaneamento
    
    Para cada (posicao, caractere) obtidos ao enumerar a mensagem_codificada:
        Se caractere for igual a 'R':
            Escrever mensagem de ruído ignorado
            Pular para a próxima iteração (continue)
            
        Se caractere for igual a 'H':
            Escrever alerta de ameaça detectada
            Interromper o laço de repetição (break)
            
        Escrever mensagem de sinal processado com sucesso
        
    Escrever mensagem de conclusão do processamento
```

````python
def escanear_transmissao_cripto(mensagem_codificada):
    """
    Escaneia caracteres de uma transmissão acústica, aplicando filtros e paradas.
    """
    print(f"Iniciando escaneamento tático da transmissão: '{mensagem_codificada}'\n")
    
    for posicao, caractere in enumerate(mensagem_codificada):
        if caractere == 'R':
            print(f"Posição #{posicao}: [!] Ruído de fundo detectado. Ignorando e continuando...")
            continue
            
        if caractere == 'H':
            print(f"Posição #{posicao}: 🚨 AMEAÇA HOSTIL DETECTADA! Bloqueando console e abortando escaneamento.")
            break
            
        print(f"Posição #{posicao}: [✓] Sinal '{caractere}' processado com sucesso.")
        
    print("\n--- Processamento de Transmissão Finalizado ---\n")

# Caso 1: Transmissão com ruídos ('R') e uma ameaça ('H') no meio
escanear_transmissao_cripto("SRSSRHS")
````

## 3) Introdução às Listas e Suas Principais Funções

Até agora, cada variável que declaramos armazenava apenas um único valor por vez. Em problemas reais de engenharia e análise de dados, precisamos trabalhar com conjuntos de dados (como uma série temporal de temperaturas de sensores ou nomes de oficiais de serviço).

Para isso, o Python nos fornece a estrutura de dados chamada **Lista** (`list`).

### O que é uma Lista?
Uma lista é uma coleção ordenada, mutável (pode ser alterada) e que permite elementos duplicados. Em Python, as listas são representadas por colchetes `[]` e seus elementos são separados por vírgula.

### Principais Funções e Métodos de Listas

O Python oferece diversas funções integradas e métodos internos para manipular listas. Veja a tabela resumo a seguir:

| Função / Método | Descrição | Exemplo |
| :--- | :--- | :--- |
| `len(lista)` | Retorna a quantidade de elementos na lista (função built-in). | `len([10, 20]) -> 2` |
| `.append(item)` | Adiciona um elemento ao **final** da lista. | `lista.append(50)` |
| `.insert(indice, item)`| Insere um elemento em uma posição (índice) específica. | `lista.insert(0, 10)` |
| `.remove(item)` | Remove a primeira ocorrência do elemento especificado. | `lista.remove(20)` |
| `.pop(indice)` | Remove e retorna o item na posição especificada (por padrão, o último). | `lista.pop()` |
| `.index(item)` | Retorna o índice (posição) da primeira ocorrência do item. | `lista.index(10)` |
| `.count(item)` | Retorna a quantidade de ocorrências de um item na lista. | `lista.count(5)` |
| `.sort()` | Ordena a lista em ordem crescente (in-place). | `lista.sort()` |
| `.reverse()` | Inverte a ordem dos elementos da lista (in-place). | `lista.reverse()` |
| `.clear()` | Remove todos os elementos da lista, deixando-a vazia. | `lista.clear()` |

## 4) Exemplos Básicos: Declarações e Indexação

Vamos aprender como criar listas em Python e acessar seus elementos individualmente por meio de seus índices.

**Regras de Acesso e Índices**:
*   O primeiro elemento da lista está no índice `0`.
*   O último elemento da lista pode ser acessado pelo índice `-1` (indexação negativa, muito útil para ler do fim para o começo).
*   Podemos extrair subpartes de uma lista usando a técnica de fatiamento (*slicing*): `lista[inicio:fim]`. Note que o índice de `fim` não é incluído na extração.

````python
# 1. Declaração de listas
lista_vazia = []
frequencias_sonar = [150, 320, 480, 210]  # Lista de inteiros
tripulacao = ["Comandante Silva", "Tenente Lucas", "Sargento Mariana"]  # Lista de strings
dados_mistos = [3.14, "Reator OK", True, 42]  # Lista com múltiplos tipos

# Exibindo as locais criadas
print("Lista Vazia:", lista_vazia)
print("Frequências do Sonar:", frequencias_sonar)
print("Tripulação de Bordo:", tripulacao)
print("Dados Mistos do Sensor:", dados_mistos)
print("-" * 50)

# 2. Indexação e acesso a elementos
primeiro_tripulante = tripulacao[0]
ultimo_tripulante = tripulacao[-1]   # Equivalente a tripulacao[2]

print("Primeiro oficial da lista:", primeiro_tripulante)
print("Último oficial da lista:", ultimo_tripulante)

# 3. Modificação de elementos
print("\nTripulação antes da substituição:", tripulacao)
tripulacao[1] = "Tenente Roberto"  # Substitui o elemento no índice 1
print("Tripulação após a substituição:", tripulacao)
print("-" * 50)

# 4. Fatiamento (Slicing)
# Extraindo as frequências intermediárias (índices 1 e 2)
sub_frequencias = frequencias_sonar[1:3]
print("Frequências originais:", frequencias_sonar)
print("Fatiamento [1:3] (índices 1 e 2):", sub_frequencias)
````

### Demonstração de Métodos em Ação

Vejamos como utilizar na prática os métodos de inserção, remoção e ordenação que vimos na tabela teórica.

````python
# Inicializando uma lista de alvos identificados pelo sonar
alvos_detectados = ["Submarino 1", "Fragata 1"]
print("Lista inicial de alvos:", alvos_detectados)

# Adicionando um novo elemento no fim (append)
alvos_detectados.append("Destróier 1")
print("Após append:", alvos_detectados)

# Inserindo um elemento numa posição específica (insert no índice 1)
alvos_detectados.insert(1, "Porta-Aviões 1")
print("Após insert no índice 1:", alvos_detectados)

# Removendo um elemento específico pelo valor (remove)
alvos_detectados.remove("Fragata 1")
print("Após remover 'Fragata 1':", alvos_detectados)

# Removendo e resgatando o último elemento (pop)
ultimo_alvo = alvos_detectados.pop()
print(f"Alvo removido com pop(): {ultimo_alvo}")
print("Lista após o pop:", alvos_detectados)

# Ordenando a lista alfabeticamente (sort)
alvos_detectados.sort()
print("Após sort (ordenação):", alvos_detectados)

# Invertendo a ordem atual da lista (reverse)
alvos_detectados.reverse()
print("Após reverse (inversão):", alvos_detectados)

# Tamanho final da lista
print(f"Quantidade final de alvos monitorados: {len(alvos_detectados)}")
````

## 5) Exemplos de Funções com Uso de Listas

Integrar listas com funções é uma das práticas mais comuns no desenvolvimento de softwares. Podemos passar uma lista como parâmetro para uma função calcular médias estatísticas, filtrar dados ou alterar elementos de coleções.

Vejamos dois exemplos práticos envolvendo dados simulados do sonar.

#### 📝 Roteiro Lógico - Exemplo A (Calcular Média)
```text
Função calcular_media_ruido(leituras_decibeis):
    Se o tamanho da lista leituras_decibeis for zero:
        Retornar 0.0
        
    Inicializar soma_total = 0.0
    Para cada leitura na lista leituras_decibeis:
        Acumular leitura em soma_total
        
    Calcular media = soma_total / tamanho da lista leituras_decibeis
    Retornar media arredondada com 2 casas decimais
```

#### 📝 Roteiro Lógico - Exemplo B (Filtrar Elementos)
```text
Função filtrar_obstaculos_proximos(lista_distancias, limite_alerta):
    Inicializar lista obstaculos_criticos como vazia
    
    Para cada dist na lista lista_distancias:
        Se dist < limite_alerta:
            Adicionar dist à lista obstaculos_criticos
            
    Retornar a lista obstaculos_criticos
```

````python
# Exemplo A: Calcular Média de Amplitude de Ruído de Fundo
def calcular_media_ruido(leituras_decibeis):
    """
    Calcula e retorna a média aritmética de uma lista de leituras acústicas.
    """
    # Evita divisão por zero caso a lista esteja vazia
    if len(leituras_decibeis) == 0:
        return 0.0
        
    soma_total = 0.0
    for leitura in leituras_decibeis:
        soma_total += leitura
        
    media = soma_total / len(leituras_decibeis)
    return round(media, 2)

# Conjunto de medições em decibéis captadas pelo sonar passivo
medicoes = [45.2, 48.7, 51.1, 46.3, 49.0]
media_db = calcular_media_ruido(medicoes)
print(f"Medições coletadas: {medicoes}")
print(f"Média do ruído de fundo: {media_db} dB\n")


# Exemplo B: Filtrar Distâncias de Colisão Imediata
def filtrar_obstaculos_proximos(lista_distancias, limite_alerta):
    """
    Itera sobre uma lista de distâncias e retorna uma nova lista
    contendo apenas os obstáculos abaixo do limite de segurança.
    """
    obstaculos_criticos = []
    
    for dist in lista_distancias:
        if dist < limite_alerta:
            # Adiciona à lista de alerta
            obstaculos_criticos.append(dist)
            
    return obstaculos_criticos

# Distâncias em metros de 6 objetos detectados pelo sonar ativo
distancias_objetos = [1200.0, 450.0, 2300.0, 150.0, 850.0, 95.0]
limite_seguranca_m = 500.0

perigos_detectados = filtrar_obstaculos_proximos(distancias_objetos, limite_seguranca_m)
print(f"Obstáculos rastreados (m): {distancias_objetos}")
print(f"Obstáculos críticos a menos de {limite_seguranca_m}m: {perigos_detectados}")
````

### Exercício 2: Somador de Amplitudes Elevadas

**Objetivo**: Iterar por uma lista numérica e aplicar condicionais para acumular e filtrar elementos específicos.

**Instruções**:
Escreva uma função chamada `somar_amplitudes_altas(lista_amplitudes, limite_minimo)` que:
1.  Inicialize um acumulador chamado `soma` em 0.0.
2.  Use um loop `for` para percorrer a `lista_amplitudes`.
3.  Verifique se o valor da amplitude é maior ou igual ao `limite_minimo`.
4.  Caso seja, adicione o valor ao acumulador `soma`.
5.  Retorne o valor acumulado da `soma` arredondado para uma casa decimal com `round(soma, 1)`.

````python
# IMPLEMENTAÇÃO DO EXERCÍCIO 2

def somar_amplitudes_altas(lista_amplitudes, limite_minimo):
    # Escreva sua lógica aqui!
    soma = 0.0
    for amplitude in lista_amplitudes:
        if amplitude >= limite_minimo:
            soma += amplitude
    return round(soma, 1)

# Lista de teste de amplitudes capturadas
amplitudes_teste = [12.5, 45.0, 78.2, 5.0, 92.1, 30.4]
limite_calibracao = 40.0

resultado_soma = somar_amplitudes_altas(amplitudes_teste, limite_calibracao)
print(f"Soma das amplitudes maiores ou iguais a {limite_calibracao} dB: {resultado_soma} dB")
# Resultado esperado: 45.0 + 78.2 + 92.1 = 215.3 dB
````

### --- Exercícios Aplicados à Missão Submarina CIAA-LPS ---

### Exercício 3: Detector de Frequências Anômalas do Sonar

**História e Enredo**:
Durante a varredura tática, o receptor acústico do submarino capta uma lista de frequências sonoras ambientais. Em condições normais de navegação, a frequência oceânica segura deve permanecer estritamente no intervalo entre $100\text{ Hz}$ e $500\text{ Hz}$ (inclusive). Qualquer sinal com frequência fora dessa faixa é classificado como uma "anomalia acústica" (que pode indicar contatos biológicos de baixíssima frequência ou motores hostis em rotação muito alta).

**Instruções**:
Escreva a função `detectar_anomalias_acusticas(lista_frequencias)` que:
1.  Crie uma lista vazia chamada `anomalias`.
2.  Itere sobre cada frequência contida na `lista_frequencias` recebida.
3.  Verifique se a frequência é menor que $100\text{ Hz}$ ou maior que $500\text{ Hz}$.
4.  Caso a frequência atenda ao critério de anomalia, adicione-a à lista `anomalias`.
5.  Retorne a lista `anomalias` ordenada de forma crescente.

````python
# IMPLEMENTAÇÃO DO EXERCÍCIO 3

def detectar_anomalias_acusticas(lista_frequencias):
    # Escreva sua lógica aqui!
    anomalias = []
    for freq in lista_frequencias:
        if freq < 100 or freq > 500:
            anomalias.append(freq)
    anomalias.sort()
    return anomalias

# Frequências captadas na varredura do setor norte
frequencias_captadas = [320, 85, 450, 620, 150, 50, 480, 510]
lista_alertas = detectar_anomalias_acusticas(frequencias_captadas)

print("Frequências anômalas identificadas (ordenadas):", lista_alertas)
# Resultado esperado: [50, 85, 510, 620]
````

### Exercício 4: Classificador Tático de Contatos Acústicos Multidimensionais

**História e Enredo**:
O computador tático do submarino detecta múltiplos contatos simultâneos na tela do sonar. Cada contato é representado por uma lista contendo três dados numéricos: `[frequencia, amplitude, tempo_eco]`, onde:
*   `frequencia`: Frequência da leitura acústica em Hz.
*   `amplitude`: Amplitude do sinal em dB.
*   `tempo_eco`: Tempo que o eco demorou para ir e voltar do alvo em segundos.

Você deve reutilizar a lógica de classificação de alvos que desenvolvemos na primeira e segunda aula, porém agora aplicando-a de forma automatizada sobre múltiplos alvos recebidos em uma lista de contatos.

**Instruções**:
1.  Reescreva a função auxiliar `calcular_distancia_aux(velocidade_som, tempo_eco)` que retorna a distância simples em metros utilizando a fórmula:
    $$\text{distancia} = \frac{\text{velocidade\_som} \times \text{tempo\_eco}}{2}$$
2.  Escreva a função principal `classificar_lista_contatos(lista_contatos, velocidade_som)` que:
    *   Crie uma lista vazia chamada `classificacoes`.
    *   Itere sobre cada contato contido na `lista_contatos` (cada contato é uma sublista contendo `[frequencia, amplitude, tempo_eco]`).
    *   Calcule a distância até o contato chamando a função `calcular_distancia_aux` com `velocidade_som` e o `tempo_eco` correspondente.
    *   Classifique o alvo seguindo as regras:
        *   Se a frequência for entre $500\text{ Hz}$ e $1500\text{ Hz}$ (inclusive) AND amplitude maior ou igual a $80\text{ dB}$: classifica como `"Submarino"`.
        *   Se a frequência for entre $20\text{ Hz}$ e $200\text{ Hz}$ (inclusive) AND amplitude maior ou igual a $60\text{ dB}$: classifica como `"Baleia"`.
        *   Caso contrário: classifica como `"Ruído"`.
    *   Adicione a string com a classificação correspondente à lista `classificacoes`.
    *   Ao final do processamento de todos os alvos, retorne a lista `classificacoes`.

````python
# IMPLEMENTAÇÃO DO EXERCÍCIO 4

def calcular_distancia_aux(velocidade_som, tempo_eco):
    # Escreva sua lógica aqui!
    return (velocidade_som * tempo_eco) / 2

def classificar_lista_contatos(lista_contatos, velocidade_som):
    # Escreva sua lógica aqui!
    classificacoes = []
    for contato in lista_contatos:
        frequencia = contato[0]
        amplitude = contato[1]
        tempo_eco = contato[2]
        
        dist = calcular_distancia_aux(velocidade_som, tempo_eco)
        
        if 500 <= frequencia <= 1500 and amplitude >= 80:
            classificacao = "Submarino"
        elif 20 <= frequencia <= 200 and amplitude >= 60:
            classificacao = "Baleia"
        else:
            classificacao = "Ruído"
            
        classificacoes.append(classificacao)
    return classificacoes

# Banco de contatos captados pelo sonar ativo
# Formato do contato: [frequencia_hz, amplitude_db, tempo_eco_s]
contatos_banco = [
    [850.0, 95.0, 1.4],   # Esperado: Submarino
    [50.0, 72.0, 1.2],    # Esperado: Baleia
    [1200.0, 45.0, 2.0],  # Esperado: Ruído (frequência bate, mas amplitude baixa)
    [10.0, 85.0, 0.5]     # Esperado: Ruído (frequência fora das faixas)
]
velocidade_som_agua = 1500.0 # m/s

diagnosticos = classificar_lista_contatos(contatos_banco, velocidade_som_agua)
print("Relatório de Classificação Tática de Alvos:")
for i, classe in enumerate(diagnosticos):
    print(f"Contato #{i+1}: Classificado como '{classe}'")
````
