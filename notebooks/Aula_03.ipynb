## 1) Exemplo Prático: Loops for, if/else e Funções

Antes de avançarmos para as listas, vamos consolidar a integração de loops, condicionais e funções.

Imagine que precisamos simular o monitoramento da pressão da câmara de ar comprimido para o sistema de lastro do submarino por um determinado número de ciclos. A cada ciclo de teste, realizamos uma leitura de pressão simulada. Se a leitura for superior a um limite seguro de tolerância, devemos emitir um alerta crítico. Caso contrário, reportamos que o sistema está operando estavelmente.

#### 📝 Roteiro Lógico (Pseudocódigo)
```text
Função monitorar_pressao_lastro(total_ciclos, limite_segurança):
    Escrever mensagem de início de monitoramento
    
    Para ciclo variando de 1 até total_ciclos (inclusive):
        pressao_lida = 1.0 + (ciclo * 0.05) - (resto de (ciclo / 3) * 0.08)
        
        Se pressao_lida > limite_segurança:
            Escrever alerta de pressão crítica
        Senão:
            Escrever log de pressão estável
            
    Escrever mensagem de conclusão
```

````python
def monitorar_pressao_lastro(total_ciclos, limite_segurança):
    """
    Simula o monitoramento da pressão de ar de lastro por N ciclos e emite alertas.
    """
    print(f"=== INICIANDO MONITORAMENTO DE SEGURANÇA ({total_ciclos} ciclos) ===")
    
    for ciclo in range(1, total_ciclos + 1):
        # Simulação matemática da pressão variando em função do ciclo
        pressao_lida = 1.0 + (ciclo * 0.05) - (ciclo % 3 * 0.08)
        
        if pressao_lida > limite_segurança:
            print(f"Ciclo #{ciclo}: Pressão = {pressao_lida:.2f} atm | ⚠️ ALERTA: Pressão crítica excedida!")
        else:
            print(f"Ciclo #{ciclo}: Pressão = {pressao_lida:.2f} atm | [OK] Pressão estável.")
            
    print("=== MONITORAMENTO CONCLUÍDO ===\n")

# Executando a simulação de monitoramento para 6 ciclos com limite de 1.15 atm
monitorar_pressao_lastro(total_ciclos=6, limite_segurança=1.15)
````

## 2) Exemplo Prático: Iteração Avançada com enumerate(), break e continue

A função integrada `enumerate()` é ideal para quando precisamos iterar sobre uma sequência (como os caracteres de uma mensagem codificada) e obter simultaneamente a **posição (índice)** e o **valor** do elemento atual.

Para controlar o comportamento do loop, relembramos o uso de:
*   `continue`: Interrompe a iteração atual do loop e pula imediatamente para a próxima.
*   `break`: Interrompe o loop por completo, saindo de sua execução imediatamente.

**Exemplo Aplicado**:
O operador do rádio do submarino interceptou uma cadeia de sinais codificados como uma string de texto. Cada caractere representa uma leitura acústica específica:
*   `'S'` (Safe): Sinal seguro (deve ser processado).
*   `'R'` (Ruído): Ruído de fundo (deve ser ignorado usando `continue`).
*   `'H'` (Hostile): Sinal de ameaça hostil detectado (o escaneamento deve ser imediatamente abortado por segurança usando `break`).

#### 📝 Roteiro Lógico (Pseudocódigo)
```text
Função escanear_transmissao_cripto(mensagem_codificada):
    Escrever mensagem de início de escaneamento
    
    Para cada (posicao, caractere) obtidos ao enumerar a mensagem_codificada:
        Se caractere for igual a 'R':
            Escrever mensagem de ruído ignorado
            Pular para a próxima iteração (continue)
            
        Se caractere for igual a 'H':
            Escrever alerta de ameaça detectada
            Interromper o laço de repetição (break)
            
        Escrever mensagem de sinal processado com sucesso
        
    Escrever mensagem de conclusão do processamento
```

````python
def escanear_transmissao_cripto(mensagem_codificada):
    """
    Escaneia caracteres de uma transmissão acústica, aplicando filtros e paradas.
    """
    print(f"Iniciando escaneamento tático da transmissão: '{mensagem_codificada}'\n")
    
    for posicao, caractere in enumerate(mensagem_codificada):
        if caractere == 'R':
            print(f"Posição #{posicao}: [!] Ruído de fundo detectado. Ignorando e continuando...")
            continue
            
        if caractere == 'H':
            print(f"Posição #{posicao}: 🚨 AMEAÇA HOSTIL DETECTADA! Bloqueando console e abortando escaneamento.")
            break
            
        print(f"Posição #{posicao}: [✓] Sinal '{caractere}' processado com sucesso.")
        
    print("\n--- Processamento de Transmissão Finalizado ---\n")

# Caso 1: Transmissão com ruídos ('R') e uma ameaça ('H') no meio
escanear_transmissao_cripto("SRSSRHS")
````
