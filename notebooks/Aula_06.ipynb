## Introdução a Dicionários

Um dicionário é uma coleção ordenada (preservando a ordem de inserção desde o Python 3.7), mutável e que **não permite chaves duplicadas**. Ao contrário de listas, que são acessadas por índices numéricos sequenciais (`0`, `1`, `2`), os dicionários utilizam **Chaves (`keys`)** customizadas (geralmente strings ou números) para mapear e recuperar **Valores (`values`)**.

Podemos visualizar um dicionário como uma tabela de consulta rápida, onde cada linha possui uma chave única e seu respectivo valor associado. Para declarar um dicionário, utilizamos chaves `{}` e separamos cada par por dois pontos `:`.

```python
# 1. Declaração de um dicionário (dados de um sensor do submarino)
dados_sensor = {
    "id": "SNS-992",
    "tipo": "Temperatura",
    "valor": 45.8,
    "unidade": "°C",
    "status": "Operacional"
}

print("Dicionário completo:", dados_sensor)
```

```python
# 2. Acesso aos valores usando colchetes [chave]
sensor_id = dados_sensor["id"]
sensor_valor = dados_sensor["valor"]

print(f"Sensor ID: {sensor_id}")
print(f"Temperatura lida: {sensor_valor} {dados_sensor['unidade']}")
```

```python
# 3. Adicionando e modificando elementos no dicionário
# Modificando o valor de uma chave existente
dados_sensor["valor"] = 48.2

# Adicionando uma nova chave-valor (data da última calibração)
dados_sensor["ultima_calibracao"] = "2026-06-01"

print("Dicionário modificado:", dados_sensor)
```

## Tabela de Principais Métodos e Funções

Veja a tabela resumo com as principais funções e métodos internos para manipulação de dicionários:

| Função / Método | Descrição | Exemplo |
| :--- | :--- | :--- |
| `len(dict)` | Retorna a quantidade de pares chave-valor no dicionário. | `len({"a": 1}) -> 1` |
| `.keys()` | Retorna uma coleção contendo todas as chaves do dicionário. | `dict.keys()` |
| `.values()` | Retorna uma coleção contendo todos os valores do dicionário. | `dict.values()` |
| `.items()` | Retorna uma coleção de tuplas `(chave, valor)`. | `dict.items()` |
| `.get(chave, default)`| Acessa o valor de uma chave com segurança. Retorna `default` se a chave não existir. | `dict.get("id", "N/A")` |
| `.update(outro)` | Atualiza ou insere múltiplos pares chave-valor a partir de outro dicionário. | `dict.update({"valor": 50})` |
| `.pop(chave)` | Remove e retorna o valor associado à chave especificada. | `val = dict.pop("status")` |
| `del dict[chave]` | Remove a chave e seu valor associado do dicionário. | `del dict["unidade"]` |
| `.clear()` | Remove todos os elementos do dicionário, deixando-a vazia. | `dict.clear()` |

```python
# 1. Recuperando chaves e valores individuais
chaves = dados_sensor.keys()
valores = dados_sensor.values()

print("Todas as Chaves:", list(chaves))
print("Todos os Valores:", list(valores))
```

```python
# 2. Recuperando pares chave-valor como tuplas usando .items()
itens = dados_sensor.items()
print("Itens (pares chave-valor):", list(itens))
```

```python
# 2. Uso do método .get() para evitar KeyErrors
# Se tentarmos acessar uma chave que não existe usando colchetes, o Python gerará um erro.
# O método .get() retorna None ou um valor padrão (default) caso a chave não seja encontrada.

# Chave que não existe
localizacao = dados_sensor.get("localizacao", "Indefinida")
print("Localização do sensor:", localizacao)

# Chave que existe
tipo_sensor = dados_sensor.get("tipo", "Não informado")
print("Tipo do sensor:", tipo_sensor)
```

```python
# 1. Remoção de chaves usando o método .pop() (retorna o valor removido)
# Vamos remover a chave "ultima_calibracao" adicionada anteriormente
valor_removido = dados_sensor.pop("ultima_calibracao")
print(f"Data de calibração removida: {valor_removido}")
```

```python
# 2. Remoção de chaves usando a palavra-chave del
# Remove diretamente a chave "status" sem retornar valor
del dados_sensor["status"]
print("Dicionário após remoções:", dados_sensor)
```

```python
# 1. Iterando apenas pelas chaves do dicionário
print("--- Chaves do Dicionário ---")
for chave in dados_sensor:
    print(chave)
```

```python
# 2. Iterando por chaves e valores simultaneamente usando .items()
print("--- Detalhes do Sensor ---")
for chave, valor in dados_sensor.items():
    print(f"{chave.upper()}: {valor}")
```

## Diferença de Performance em Buscas: Lista vs Dicionário

Uma das maiores vantagens dos dicionários sobre as listas é a **eficiência em operações de busca**. 

*   **Listas**: O Python precisa percorrer elemento por elemento a partir do início para encontrar o valor buscado. Em computação, chamamos isso de busca linear com complexidade de tempo **$O(N)$**. Se a lista tem 20.000 elementos, o pior caso exige 20.000 comparações!
*   **Dicionários**: O Python utiliza uma estrutura interna baseada em **Tabela Hash**. Isso permite calcular matematicamente o endereço de memória associado a uma chave e acessá-lo instantaneamente, independentemente do tamanho do dicionário. Chamamos isso de busca em tempo constante com complexidade **$O(1)$**!

Vamos simular isso criando uma base de dados de contatos do sonar com 20.000 registros e comparar os tempos de busca de uma ID específica.

```python
import time

# Criando uma base de dados simulada de contatos acústicos com 20.000 registros
total_registros = 20000

# 1. Base de dados como Lista de Dicionários (O(N) busca linear)
lista_contatos = []
for i in range(total_registros):
    lista_contatos.append({"id": f"ALVO-{i}", "tipo": "Desconhecido", "distancia": 500 + i})

# 2. Base de dados como Dicionário de Dicionários (O(1) busca direta)
dict_contatos = {}
for i in range(total_registros):
    dict_contatos[f"ALVO-{i}"] = {"tipo": "Desconhecido", "distancia": 500 + i}

# Definimos o alvo procurado como o último elemento (pior cenário para a busca linear)
target_id = f"ALVO-{total_registros - 1}"
print(f"Base de dados criada com {total_registros} registros. Alvo de busca: {target_id}")
```

```python
# --- BUSCA NA LISTA (Busca linear O(N)) ---
# O Python precisa ler item por item até achar o ID correspondente
inicio_lista = time.time()
alvo_encontrado_lista = None
for contato in lista_contatos:
    if contato["id"] == target_id:
        alvo_encontrado_lista = contato
        break
fim_lista = time.time()
tempo_lista = (fim_lista - inicio_lista) * 1000 # em milissegundos

print(f"Busca na Lista: {tempo_lista:.6f} ms")
print("Alvo encontrado na lista:", alvo_encontrado_lista)
```

```python
# --- BUSCA NO DICIONÁRIO (Busca direta por chave O(1)) ---
# O Python consulta o hash da chave e vai direto no endereço de memória
inicio_dict = time.time()
alvo_encontrado_dict = dict_contatos.get(target_id)
fim_dict = time.time()
tempo_dict = (fim_dict - inicio_dict) * 1000 # em milissegundos

print(f"Busca no Dicionário: {tempo_dict:.6f} ms")
print("Alvo encontrado no dicionário:", alvo_encontrado_dict)
print(f"--> O dicionário foi cerca de {tempo_lista / max(tempo_dict, 0.000001):.1f} vezes mais rápido!")
```

## Funções com Estruturas Aninhadas (Banco de Dados em Memória)

Na programação prática, frequentemente combinamos listas e dicionários para modelar bancos de dados em memória. As duas abordagens mais comuns são:
1. **Lista de Dicionários**: Ideal para quando a ordem dos elementos importa ou quando precisamos realizar varreduras e filtros sobre toda a coleção.
2. **Dicionário de Dicionários**: Ideal para quando precisamos de acesso direto e instantâneo aos registros através de um identificador único (chave primária).

Vejamos como escrever funções que recebem e processam essas estruturas aninhadas.

```python
# Exemplo A: Função que processa uma LISTA DE DICIONÁRIOS
# Banco de dados simulado contendo dicionários de alvos
sonar_alvos = [
    {"id": "ALVO-01", "tipo": "Submarino", "distancia": 450.0},
    {"id": "ALVO-02", "tipo": "Baleia", "distancia": 1200.0},
    {"id": "ALVO-03", "tipo": "Navio", "distancia": 250.0},
]

def filtrar_alvos_proximos(lista_alvos, limite_distancia):
    """
    Varre a lista de dicionários e retorna apenas os alvos
    com distância menor ou igual à distância limite.
    """
    alvos_criticos = []
    for alvo in lista_alvos:
        # Acessamos a chave 'distancia' de cada dicionário na lista
        if alvo["distancia"] <= limite_distancia:
            alvos_criticos.append(alvo)
    return alvos_criticos

# Chamando a função para encontrar alvos a menos de 500 metros
alertas = filtrar_alvos_proximos(sonar_alvos, 500.0)
print("Alvos críticos detectados (Menos de 500m):")
for item in alertas:
    print(f"- {item['tipo']} ({item['id']}) a {item['distancia']}m")
```

```python
# Exemplo B: Função que processa um DICIONÁRIO DE DICIONÁRIOS
# O banco de dados mapeia a ID do alvo direto para o seu dicionário de detalhes
banco_dados_alvos = {
    "ALVO-01": {"tipo": "Submarino", "distancia": 450.0, "status": "Suspeito"},
    "ALVO-02": {"tipo": "Baleia", "distancia": 1200.0, "status": "Neutro"},
    "ALVO-03": {"tipo": "Navio", "distancia": 250.0, "status": "Hostil"}
}

def atualizar_status_alvo(db_dicionario, alvo_id, novo_status):
    """
    Busca o alvo no dicionário pelo ID e atualiza seu status de forma segura.
    """
    # Buscamos o dicionário interno correspondente à chave alvo_id
    alvo = db_dicionario.get(alvo_id)
    
    if alvo:
        # Modificamos o dicionário interno
        alvo["status"] = novo_status
        print(f"[OK] Status do {alvo_id} atualizado para: {novo_status}")
    else:
        print(f"[!] Erro: Alvo {alvo_id} não cadastrado no banco de dados.")

# Atualizando o status do ALVO-01 de 'Suspeito' para 'Hostil'
atualizar_status_alvo(banco_dados_alvos, "ALVO-01", "Hostil")
print("Dados atualizados do ALVO-01:", banco_dados_alvos["ALVO-01"])
```
